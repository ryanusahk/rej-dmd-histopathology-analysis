```python
def auto_pair(image_files):
    if len(image_files)%2 != 0:
        print('ERROR')
        return
    return [(a, b) for a, b in zip(image_files[:(len(image_files)//2)], image_files[(len(image_files)//2):])]
```

```python
def rescale(img, f):
    imgc = img.copy().astype('uint16')
    imgc = imgc*f
    imgc[imgc>255] = 255
    imgc = imgc.astype('uint8')
    return imgc
```

```python
def create_rgb_like(ref):
    if len(ref.shape) == 3:
        return np.zeros_like(ref).astype('uint8')
    else:
        return np.zeros((ref.shape[0], ref.shape[1], 3)).astype('uint8')

def add_rgb(base, gray, channel):
    grgb = create_rgb_like(gray)
    grgb[:,:,channel] = gray
    return np.maximum(base, grgb)

def show_image(img):
    plt.subplots(figsize=(11,11))
    plt.imshow(img)
    plt.show()
```

```python
muscle_sections = {}

for excel in tqdm.tqdm(forms[:]):
    
    rt = excel.split('/')[-1].split('_form')[0]
    
    try:
        img = pickle.load(open(TARGET_FOLDER+'/'+rt+'_fl.pickle', 'rb'))
    except:
        print('failed to load image pickle')
        print(excel)
        continue
    valid_contours = pickle.load(open(TARGET_FOLDER+'/'+rt+'_contours.pickle', 'rb'))

    excel_df = pd.read_excel(excel, engine='openpyxl')

    keynames = []

    for i in excel_df['contour_id']:
        keynames.append(rt+'.'+str(i))

    for ii, (index, c) in enumerate(zip(keynames, valid_contours)):
            #     temp = (img[:,:,1].copy()//256).astype('uint8')
        x, y, w, h = cv.boundingRect(c)

        muscle_sections[index] = {}
        muscle_sections[index]['section'] = {}

        muscle_sections[index]['section']['dystrophin'] = img[y:y+h, x:x+w, DYST]
        muscle_sections[index]['section']['auto'] = img[y:y+h, x:x+w, AUTO]
        muscle_sections[index]['section']['dapi'] = img[y:y+h, x:x+w, DAPI]
        muscle_sections[index]['section']['lama'] = img[y:y+h, x:x+w, LAMA]
        muscle_sections[index]['section']['mouse']= excel_df['mouse_num'][ii]
        muscle_sections[index]['section']['descriptor'] = excel_df['descriptor'][ii]

        mask = cv.drawContours(image=np.zeros_like(img[:, :, 1]), contours=[
                               c], contourIdx=-1, color=(255), thickness=-1)
        mask255 = mask[y:y+h, x:x+w]
        dist_map = cv.distanceTransform(mask255, cv.DIST_L2, cv.DIST_MASK_PRECISE)
        _, radius, _, center = cv.minMaxLoc(dist_map)

        unit = int(0.95*radius/np.sqrt(2))
        y = center[0]
        x = center[1]

        muscle_sections[index]['roi'] = {}
        muscle_sections[index]['roi']['dystrophin'] = muscle_sections[index]['section']['dystrophin'][x -
                                                                                                      unit:x+unit, y-unit:y+unit]
        muscle_sections[index]['roi']['auto'] = muscle_sections[index]['section']['auto'][x -
                                                                                          unit:x+unit, y-unit:y+unit]
        muscle_sections[index]['roi']['dapi'] = muscle_sections[index]['section']['dapi'][x -
                                                                                          unit:x+unit, y-unit:y+unit]
        muscle_sections[index]['roi']['lama'] = muscle_sections[index]['section']['lama'][x -
                                                                                          unit:x+unit, y-unit:y+unit]

        th3 = cv.adaptiveThreshold(
            muscle_sections[index]['roi']['lama'], 255, cv.ADAPTIVE_THRESH_GAUSSIAN_C, cv.THRESH_BINARY,311, .5)
        th3_bin = th3.copy()

        cell_contours, cell_hierarchy = cv.findContours(
            image=th3_bin, mode=cv.RETR_TREE, method=cv.CHAIN_APPROX_SIMPLE)

        valid_cell_contours = []
        circularities = []
        contour_areas = [cv.contourArea(c) for c in cell_contours]

        mode_ch = sp.stats.mode(cell_hierarchy[0][:, 3])[0][0]
        for cc, ca, ch in zip(cell_contours, contour_areas, cell_hierarchy[0]):
            #         if ch[3] == mode_ch and ca > 200 and ca < 2000:
            if ch[3] == mode_ch and ca > 50:

                perimeter = cv.arcLength(cc, True)

                circularity = 4*np.pi*ca/(perimeter**2)

                if circularity < 0.15:
                    continue

                valid_cell_contours.append(cc)
                circularities.append(circularity)

        muscle_sections[index]['segmented_cells'] = {}
        muscle_sections[index]['segmented_cells']['contours'] = valid_cell_contours
        muscle_sections[index]['segmented_cells']['circularity'] = circularities

        segmented_cell_image = create_rgb_like(
            muscle_sections[index]['roi']['dystrophin'])
    #     segmented_cell_image = add_rgb(segmented_cell_image, rescale(muscle_sections[index]['roi']['lama'], 25),2)
    #     segmented_cell_image = add_rgb(segmented_cell_image, th3, 1)

        for c in valid_cell_contours:
            rc = np.random.randint(100, 255, 3)
    #         rc = (rc[0],rc[1],rc[2]//3)
            random_color = (int(rc[0]), int(rc[1]), int(rc[2]))
            segmented_cell_image = cv.drawContours(image=segmented_cell_image, contours=[
                                                   c], contourIdx=-1, color=random_color, thickness=-1)

        muscle_sections[index]['segmented_cells']['preview'] = segmented_cell_image

        segmented_cell_image = create_rgb_like(
            muscle_sections[index]['roi']['dystrophin'])
        segmented_cell_image = cv.drawContours(
            image=segmented_cell_image, contours=valid_cell_contours, contourIdx=-1, color=(255, 255, 255), thickness=-1)
        muscle_sections[index]['segmented_cells']['preview2'] = segmented_cell_image    
```

```python
# Q DYST MEAN
for i in tqdm.tqdm(muscle_sections.keys()):
    dyst_means = []
    for cc in (muscle_sections[i]['segmented_cells']['contours']):
        blank = np.zeros_like(muscle_sections[i]['roi']['dapi'])
        mask255 = cv.drawContours(image=blank, contours=[cc], contourIdx=-1, color=(255), thickness=7)
        ones = np.ones_like(muscle_sections[i]['roi']['dapi'])
        body = cv.drawContours(image=ones, contours=[cc], contourIdx=-1, color=(0), thickness=-1)
        mask255 = np.multiply(mask255, body)

        mask1 = (mask255>0).astype(int)
        maskbool = mask1.astype(bool) 
        isolated_dyst = muscle_sections[i]['roi']['dystrophin'].flatten()[maskbool.flatten()]
        dyst_means.append(np.mean(isolated_dyst))
    muscle_sections[i]['segmented_cells']['dyst_mean'] = np.array(dyst_means)
```
