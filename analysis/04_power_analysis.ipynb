```python
import pandas as pd
import numpy as np
from scipy import stats
import statsmodels.api as sm
from statsmodels.genmod import families
from statsmodels.genmod.cov_struct import Exchangeable
from statsmodels.stats.anova import anova_lm

# Define binary outcome variables to check
binary_outcomes = ['central_nuclei2_bin', 'damaged_bin', 'hypertrophic_bin']

# Function to calculate ICC from GEE model
def calculate_icc_from_gee(df, outcome_col, cluster_col='filename'):
   """
   Calculate ICC for binary outcome using GEE exchangeable correlation structure
   """
   if outcome_col not in df.columns:
       return None
   
   # Remove rows with missing values for this outcome
   clean_df = df.dropna(subset=[outcome_col, cluster_col])
   
   if len(clean_df) == 0:
       return None
   
   # Create intercept-only model (null model for ICC calculation)
   y = clean_df[outcome_col]
   X = np.ones(len(clean_df))  # Intercept only
   
   # Fit GEE model with exchangeable correlation
   try:
       gee_model = sm.GEE(y, X, groups=clean_df[cluster_col], 
                          cov_struct=Exchangeable(), 
                          family=families.Binomial())
       gee_result = gee_model.fit()
       
       # Extract ICC (correlation parameter) from exchangeable structure
       # Handle case where dep_params might be a scalar instead of array
       dep_params = gee_result.cov_struct.dep_params
       if hasattr(dep_params, '__len__') and len(dep_params) > 0:
           icc = dep_params[0]
       elif isinstance(dep_params, (int, float, np.number)):
           icc = float(dep_params)
       else:
           icc = None
           
       return icc
       
   except Exception as e:
       print(f"Error calculating ICC for {outcome_col}: {e}")
       return None

# Initialize dictionary to store all ICC results
icc_results = {}

# Calculate ICC for each dataset and outcome
datasets = {'minidystrophin_df': minidystrophin_df, 'abe_df': abe_df}

print("Intracluster Correlation Coefficients (ICC) by Dataset, Treatment Group, and Outcome:")
print("=" * 85)

for dataset_name, df in datasets.items():
   print(f"\n{dataset_name.upper()}:")
   print("-" * 50)
   
   # Initialize nested dictionary for this dataset
   icc_results[dataset_name] = {}
   
   # Overall ICC (all treatment groups combined)
   print("\nOVERALL (All Treatment Groups):")
   icc_results[dataset_name]['overall'] = {}
   
   for outcome in binary_outcomes:
       if outcome in df.columns:
           icc = calculate_icc_from_gee(df, outcome)
           icc_results[dataset_name]['overall'][outcome] = icc
           if icc is not None:
               print(f"    {outcome:20s}: {icc:.4f}")
           else:
               print(f"    {outcome:20s}: Failed to calculate")
       else:
           icc_results[dataset_name]['overall'][outcome] = None
           print(f"    {outcome:20s}: Not available in dataset")
   
   # By treatment group
   treatment_groups = df['plotLabel'].unique()
   print(f"\nBY TREATMENT GROUP:")
   
   for treatment in sorted(treatment_groups):
       print(f"\n  {treatment}:")
       treatment_df = df[df['plotLabel'] == treatment]
       
       # Initialize dictionary for this treatment group
       icc_results[dataset_name][treatment] = {}
       
       for outcome in binary_outcomes:
           if outcome in df.columns:
               icc = calculate_icc_from_gee(treatment_df, outcome)
               icc_results[dataset_name][treatment][outcome] = icc
               if icc is not None:
                   print(f"      {outcome:20s}: {icc:.4f}")
               else:
                   print(f"      {outcome:20s}: Failed to calculate")
           else:
               icc_results[dataset_name][treatment][outcome] = None
               print(f"      {outcome:20s}: Not available in dataset")
       
       # Summary stats for this treatment group
       n_clusters = treatment_df['filename'].nunique()
       n_cells = len(treatment_df)
       avg_cells_per_cluster = n_cells / n_clusters if n_clusters > 0 else 0
       print(f"      Clusters: {n_clusters}, Cells: {n_cells}, Avg cells/cluster: {avg_cells_per_cluster:.1f}")
   
   # Overall dataset summary
   n_clusters = df['filename'].nunique()
   n_cells = len(df)
   avg_cells_per_cluster = n_cells / n_clusters
   print(f"\nDATASET SUMMARY:")
   print(f"    Total cells: {n_cells}")
   print(f"    Total clusters (leg sections): {n_clusters}")
   print(f"    Average cells per cluster: {avg_cells_per_cluster:.1f}")
   print(f"    Treatment groups: {sorted(treatment_groups)}")

print(f"\n" + "=" * 85)
print("ICC results stored in 'icc_results' dictionary")
print("Access pattern: icc_results[dataset][treatment_group][outcome]")
print("Example: icc_results['minidystrophin_df']['Treated']['central_nuclei2_bin']")
```

```python
import pandas as pd
import numpy as np
from scipy import stats
import statsmodels.api as sm
from statsmodels.genmod import families
from statsmodels.genmod.cov_struct import Exchangeable
from tqdm import tqdm
import matplotlib.pyplot as plt
import seaborn as sns
import pickle
import hashlib
from pathlib import Path
import itertools
from joblib import Parallel, delayed

def simulate_power_unified(baseline_prop, relative_effect, cluster_size, median_icc,
                           n_clusters=5, n_sims=100, analysis_type='clustered'):
    treatment_prop = baseline_prop * (1 - relative_effect)
    all_cluster_effects_control = np.random.normal(0, np.sqrt(median_icc), (n_sims, n_clusters))
    all_cluster_effects_treatment = np.random.normal(0, np.sqrt(median_icc), (n_sims, n_clusters))
    baseline_logit = np.log(baseline_prop / (1 - baseline_prop))
    treatment_logit = np.log(treatment_prop / (1 - treatment_prop))
    significant_results = 0
    actual_sims_completed = 0
    for sim in range(n_sims):
        cluster_effects_control = all_cluster_effects_control[sim]
        cluster_effects_treatment = all_cluster_effects_treatment[sim]
        if analysis_type == 'clustered':
            control_data, treatment_data, cluster_ids = [], [], []
            for cluster in range(n_clusters):
                cluster_logit = baseline_logit + cluster_effects_control[cluster]
                cluster_prob = np.clip(1 / (1 + np.exp(-cluster_logit)), 0.001, 0.999)
                control_data.extend(np.random.binomial(1, cluster_prob, cluster_size))
                cluster_ids.extend([f'control_{cluster}'] * cluster_size)
            for cluster in range(n_clusters):
                cluster_logit = treatment_logit + cluster_effects_treatment[cluster]
                cluster_prob = np.clip(1 / (1 + np.exp(-cluster_logit)), 0.001, 0.999)
                treatment_data.extend(np.random.binomial(1, cluster_prob, cluster_size))
                cluster_ids.extend([f'treatment_{cluster}'] * cluster_size)
            y = np.array(control_data + treatment_data)
            treatment = np.array([0] * len(control_data) + [1] * len(treatment_data))
            clusters = np.array(cluster_ids)
            try:
                X = sm.add_constant(treatment)
                gee_model = sm.GEE(y, X, groups=clusters,
                                   cov_struct=Exchangeable(),
                                   family=families.Binomial())
                gee_result = gee_model.fit(start_params=[baseline_logit, treatment_logit - baseline_logit])
                if gee_result.pvalues[1] < 0.05:
                    significant_results += 1
            except:
                pass
        else: # analysis_type == 'summary'
            control_percentages, treatment_percentages = [], []
            for cluster in range(n_clusters):
                cluster_logit = baseline_logit + cluster_effects_control[cluster]
                cluster_prob = np.clip(1 / (1 + np.exp(-cluster_logit)), 0.001, 0.999)
                cluster_outcomes = np.random.binomial(1, cluster_prob, cluster_size)
                control_percentages.append(np.mean(cluster_outcomes))
            for cluster in range(n_clusters):
                cluster_logit = treatment_logit + cluster_effects_treatment[cluster]
                cluster_prob = np.clip(1 / (1 + np.exp(-cluster_logit)), 0.001, 0.999)
                cluster_outcomes = np.random.binomial(1, cluster_prob, cluster_size)
                treatment_percentages.append(np.mean(cluster_outcomes))
            try:
                _, p_value = stats.ttest_ind(control_percentages, treatment_percentages)
                if p_value < 0.05:
                    significant_results += 1
            except:
                pass
        actual_sims_completed += 1
        if sim >= 200 and actual_sims_completed > 0:
            current_power = significant_results / actual_sims_completed
            if current_power == 0.0 or (current_power == 1.0 and sim >= 2000):
                break
    if actual_sims_completed == 0:
        return 0.0
    return min(significant_results / actual_sims_completed, 1.0)

def generate_unified_cache_key(baseline_prop, relative_effects, conditions, median_icc, n_sims):
    conditions_str = str(sorted(tuple(c.items()) for c in conditions))
    params_str = f"unified_{baseline_prop}_{relative_effects}_{conditions_str}_{median_icc}_{n_sims}"
    return hashlib.md5(params_str.encode()).hexdigest()

def save_unified_results_to_pickle(results_df, cache_key, cache_dir="power_analysis_cache"):
    cache_path = Path(cache_dir)
    cache_path.mkdir(parents=True, exist_ok=True)
    filepath = cache_path / f"unified_power_{cache_key}.pkl"
    with open(filepath, 'wb') as f:
        pickle.dump(results_df, f)
    print(f"Results saved to: {filepath}")
    return filepath

def load_unified_results_from_pickle(cache_key, cache_dir="power_analysis_cache"):
    filepath = Path(cache_dir) / f"unified_power_{cache_key}.pkl"
    if filepath.exists():
        try:
            with open(filepath, 'rb') as f:
                results_df = pickle.load(f)
            print(f"Loaded cached results from: {filepath}")
            return results_df
        except Exception as e:
            print(f"Error loading cached results: {e}")
            return None
    return None

def run_unified_power_analysis(baseline_prop=0.25, median_icc=0.015,
                               n_sims=1000, cache_dir="power_analysis_cache",
                               force_rerun=False):
    N_P_CORES = 8 # Optimal for i7-14700K
    conditions = [
        {'cells': 50, 'total_mice': 12, 'analysis': 'clustered'},
        {'cells': 100, 'total_mice': 12, 'analysis': 'clustered'},
        {'cells': 2500, 'total_mice': 12, 'analysis': 'clustered'},
        {'cells': 50, 'total_mice': 48, 'analysis': 'clustered'},
        {'cells': 100, 'total_mice': 48, 'analysis': 'clustered'},
        {'cells': 100, 'total_mice': 12, 'analysis': 'summary'},
        {'cells': 100, 'total_mice': 48, 'analysis': 'summary'},
    ]
    relative_effects = np.arange(0.05, 0.51, 0.0125)
    cache_key = generate_unified_cache_key(baseline_prop, relative_effects.tolist(),
                                           conditions, median_icc, n_sims)
    if not force_rerun:
        cached_results = load_unified_results_from_pickle(cache_key, cache_dir)
        if cached_results is not None:
            print("Using cached results - no simulation needed!")
            return cached_results
    print("No cached results found - running simulation in parallel...")
    param_combinations = list(itertools.product(conditions, relative_effects))
    print(f"Testing {len(param_combinations)} combinations across {N_P_CORES} CPU cores.")
    np.random.seed(42)
    def worker(condition, relative_effect):
        mice_per_group = condition['total_mice'] // 2
        analysis = condition['analysis']
        power = simulate_power_unified(
            baseline_prop=baseline_prop,
            relative_effect=relative_effect,
            cluster_size=condition['cells'],
            median_icc=median_icc,
            n_clusters=mice_per_group,
            n_sims=n_sims,
            analysis_type=analysis
        )
        return {
            'cells_per_mouse': condition['cells'], 'mice_per_group': mice_per_group,
            'total_mice': condition['total_mice'], 'relative_effect_pct': relative_effect * 100,
            'power': power * 100, 'analysis_type': analysis,
            'baseline_prop': baseline_prop, 'median_icc': median_icc, 'n_sims': n_sims
        }
    results = Parallel(n_jobs=N_P_CORES)(
        delayed(worker)(cond, rel_eff) for cond, rel_eff in tqdm(param_combinations, desc="Calculating power")
    )
    results_df = pd.DataFrame(results)
    save_unified_results_to_pickle(results_df, cache_key, cache_dir)
    return results_df

def create_unified_comparison_plot(results_df, figsize=(14, 7)):
    baseline_prop = results_df['baseline_prop'].iloc[0]
    median_icc = results_df['median_icc'].iloc[0]
    n_sims = results_df['n_sims'].iloc[0]

    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=figsize)
    plt.style.use('seaborn-v0_8-whitegrid')

    # Create a consistent color and style mapping for all conditions
    unique_labels = sorted(results_df['simplified_label'].unique())
    color_palette = sns.color_palette("deep", len(unique_labels))
    style_map = {
        label: {
            "color": color_palette[i],
            "linewidth": 4 if '2500 cells' in label else 2,
            "linestyle": '--' if 't-test' in label else '-'
        }
        for i, label in enumerate(unique_labels)
    }

    # --- Left Plot: GEE vs. t-test (12 total mice) ---
    ax1.set_title('GEE vs. t-test (12 total mice)')
    data_ax1 = results_df[results_df['total_mice'] == 12]
    for label in sorted(data_ax1['simplified_label'].unique()):
        subset = data_ax1[data_ax1['simplified_label'] == label]
        ax1.plot(subset['relative_effect_pct'], subset['power'],
                 label=label, **style_map[label])

    # --- Right Plot: Power Across All Conditions ---
    ax2.set_title('Power Across All Conditions')
    for label in unique_labels:
        subset = results_df[results_df['simplified_label'] == label]
        ax2.plot(subset['relative_effect_pct'], subset['power'],
                 label=label, **style_map[label])

    # --- Common Formatting ---
    for ax in [ax1, ax2]:
        ax.axhline(y=80, color='black', linestyle=':', alpha=0.7)
        ax.set_xlabel('Effect Size (% Reduction)')
        ax.set_ylim(0, 105)

    ax1.set_ylabel('Statistical Power (%)')
    ax2.set_ylabel('')
    ax1.legend(loc='lower right', fontsize=9)
    ax2.legend(bbox_to_anchor=(1.04, 1), loc='upper left', fontsize=9)
    
    plt.suptitle(f'Power Analysis: Cell-Level (GEE) vs. Mouse-Level (t-test) Analysis\n'
                 f'({baseline_prop*100}% baseline, ICC = {median_icc:.3f}, {n_sims} simulations)',
                 fontsize=14)
    plt.tight_layout(rect=[0, 0, 0.9, 0.96])
    plt.show()

def print_unified_insights(results_df):
    print("\nPower Analysis Insights:")
    print("=" * 60)
    key_effects = [20, 30, 40]
    print("\nPower comparison at 12 total mice:")
    for effect in key_effects:
        print(f"\nFor a {effect}% effect size:")
        effect_data = results_df[(results_df['relative_effect_pct'] > effect - 2) &
                                 (results_df['relative_effect_pct'] < effect + 2) &
                                 (results_df['total_mice'] == 12)]
        for cells in sorted(effect_data['cells_per_mouse'].unique()):
            clustered = effect_data[(effect_data['cells_per_mouse'] == cells) & (effect_data['analysis_type'] == 'clustered')]
            summary = effect_data[(effect_data['cells_per_mouse'] == cells) & (effect_data['analysis_type'] == 'summary')]
            if not clustered.empty:
                power_clustered = clustered['power'].iloc[0]
                if not summary.empty:
                    power_summary = summary['power'].iloc[0]
                    diff = power_clustered - power_summary
                    print(f"  {cells:<4} cells: {power_clustered:5.1f}% (GEE) vs {power_summary:5.1f}% (t-test) [+{diff:.1f}% gain]")
                else:
                    print(f"  {cells:<4} cells: {power_clustered:5.1f}% (GEE)")

    print("\nMinimum detectable effect at 80% power:")
    # Use simplified_label for this part as well
    for label in sorted(results_df['simplified_label'].unique()):
        data = results_df[results_df['simplified_label'] == label]
        above_80 = data[data['power'] >= 80]
        if not above_80.empty:
            min_effect = above_80['relative_effect_pct'].min()
            print(f"  {label:<40}: {min_effect:.1f}% effect")
        else:
            print(f"  {label:<40}: >{results_df['relative_effect_pct'].max():.1f}% effect (underpowered)")

if __name__ == "__main__":
    BASELINE_PROP = 0.25
    MEDIAN_ICC = 0.015
    N_SIMS = 10000

    unified_results_df = run_unified_power_analysis(
        baseline_prop=BASELINE_PROP,
        median_icc=MEDIAN_ICC,
        n_sims=N_SIMS,
        force_rerun=False
    )

    if not unified_results_df.empty:
        def add_simplified_label(df):
            def create_label(row):
                if row['analysis_type'] == 'summary':
                    return f"n={row['total_mice']} (t-test)"
                else:
                    return f"{row['cells_per_mouse']} cells, n={row['total_mice']}"
            df['simplified_label'] = df.apply(create_label, axis=1)
            return df
        unified_results_df = add_simplified_label(unified_results_df)
        
        print(f"\nAnalysis completed! Results stored in DataFrame with {len(unified_results_df)} rows.")
        create_unified_comparison_plot(unified_results_df)
        print_unified_insights(unified_results_df)
```

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

def interpolate_effect_size(group, target_power=80.0):
    """
    Interpolates the effect size for a given target power for a single experimental condition.
    """
    group = group.sort_values('relative_effect_pct')
    
    if group['power'].max() < target_power:
        return f"> {group['relative_effect_pct'].max():.1f}"
    
    if group['power'].min() > target_power:
        return f"< {group['relative_effect_pct'].min():.1f}"

    above_80 = group[group['power'] >= target_power].iloc[0]
    below_80 = group[group['power'] < target_power].iloc[-1]

    x1, y1 = below_80['relative_effect_pct'], below_80['power']
    x2, y2 = above_80['relative_effect_pct'], above_80['power']
    
    interpolated_effect = x1 + (target_power - y1) * (x2 - x1) / (y2 - y1)
    
    return interpolated_effect

# --- Data Preparation (Unchanged) ---
interpolated_results = unified_results_df.groupby('simplified_label').apply(interpolate_effect_size)
results_df = interpolated_results.reset_index(name='detectable_effect_at_80_power')
plot_df = results_df.copy()
plot_df['detectable_effect_at_80_power'] = pd.to_numeric(plot_df['detectable_effect_at_80_power'], errors='coerce')
plot_df.dropna(subset=['detectable_effect_at_80_power'], inplace=True)


# --- Define the color palette ---
default_color = sns.color_palette('muted')[0]
palette = {
    'n=12 (t-test)': 'gray',
    'n=48 (t-test)': 'black',
    '50 cells, n=12': default_color,
    '100 cells, n=12': default_color,
    '2500 cells, n=12': 'green',
    '50 cells, n=48': default_color,
    '100 cells, n=48': default_color
}


# --- Create the bar plot ---
plt.figure(figsize=(10, 6))
# The data is sorted by power, and we pass the palette dictionary for coloring.
# No 'order' parameter is used, so the sort order is respected.
sns.barplot(data=plot_df.sort_values(by='detectable_effect_at_80_power'), 
            x='simplified_label', 
            y='detectable_effect_at_80_power',
            palette=palette)  # Use the custom color palette

plt.title('Minimum Detectable Effect at 80% Power')
plt.ylabel('Effect Size (%)')
plt.xlabel('Experimental Condition')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

# Display the original data table as well
print(results_df)
```

```python
import pandas as pd
import numpy as np
from scipy import stats
import statsmodels.api as sm
from statsmodels.genmod import families
from statsmodels.genmod.cov_struct import Exchangeable
from tqdm import tqdm
import matplotlib.pyplot as plt
import seaborn as sns
import pickle
import hashlib
import os
from pathlib import Path

def simulate_clustered_binary_data_fixed_mice(baseline_prop, relative_effect, cluster_size, median_icc, n_clusters=5, n_sims=100):
    """
    Calculate power for fixed number of mice (clusters)
    """
    # Calculate treatment proportion from relative effect
    treatment_prop = baseline_prop * (1 - relative_effect)
    
    # Pre-generate all random effects (vectorized)
    all_cluster_effects_control = np.random.normal(0, np.sqrt(median_icc), (n_sims, n_clusters))
    all_cluster_effects_treatment = np.random.normal(0, np.sqrt(median_icc), (n_sims, n_clusters))
    
    # Pre-calculate logits
    baseline_logit = np.log(baseline_prop / (1 - baseline_prop))
    treatment_logit = np.log(treatment_prop / (1 - treatment_prop))
    
    significant_results = 0
    actual_sims_completed = 0
    
    for sim in range(n_sims):
        # Use pre-generated random effects
        cluster_effects_control = all_cluster_effects_control[sim]
        cluster_effects_treatment = all_cluster_effects_treatment[sim]
        
        control_data = []
        treatment_data = []
        cluster_ids = []
        
        # Generate control group data
        for cluster in range(n_clusters):
            cluster_logit = baseline_logit + cluster_effects_control[cluster]
            cluster_prob = 1 / (1 + np.exp(-cluster_logit))
            cluster_prob = np.clip(cluster_prob, 0.001, 0.999)
            
            cluster_outcomes = np.random.binomial(1, cluster_prob, cluster_size)
            control_data.extend(cluster_outcomes)
            cluster_ids.extend([f'control_{cluster}'] * cluster_size)
        
        # Generate treatment group data
        for cluster in range(n_clusters):
            cluster_logit = treatment_logit + cluster_effects_treatment[cluster]
            cluster_prob = 1 / (1 + np.exp(-cluster_logit))
            cluster_prob = np.clip(cluster_prob, 0.001, 0.999)
            
            cluster_outcomes = np.random.binomial(1, cluster_prob, cluster_size)
            treatment_data.extend(cluster_outcomes)
            cluster_ids.extend([f'treatment_{cluster}'] * cluster_size)
        
        # Combine data
        y = np.array(control_data + treatment_data)
        treatment = np.array([0] * len(control_data) + [1] * len(treatment_data))
        clusters = np.array(cluster_ids)
        
        # Fit GEE model
        try:
            X = sm.add_constant(treatment)
            gee_model = sm.GEE(y, X, groups=clusters, 
                              cov_struct=Exchangeable(), 
                              family=families.Binomial())
            gee_result = gee_model.fit()
            
            # Check if treatment effect is significant
            p_value = gee_result.pvalues[1]
            if p_value < 0.05:
                significant_results += 1
                
        except:
            pass
        
        actual_sims_completed += 1
        
        # Simple early stopping for extreme cases
        if sim >= 15:
            current_power = significant_results / actual_sims_completed
            if current_power == 0.0:
                break
            elif current_power == 1.0 and sim >= 20:
                break
    
    # Calculate final power
    power = significant_results / actual_sims_completed
    power = min(power, 1.0)
    return power

def find_required_mice_for_effect_size(baseline_prop, relative_effect, cluster_size, median_icc, target_power=0.8, max_mice=200, n_sims=100):
    """
    Find minimum number of mice needed for given effect size using binary search
    """
    if relative_effect == 0:  # No effect, infinite mice needed
        return max_mice
    
    low = 2
    high = max_mice
    best_estimate = None
    
    # Binary search
    while low <= high:
        mid = (low + high) // 2
        
        power = simulate_clustered_binary_data_fixed_mice(
            baseline_prop=baseline_prop,
            relative_effect=relative_effect,
            cluster_size=cluster_size,
            median_icc=median_icc,
            n_clusters=mid,
            n_sims=n_sims
        )
        
        if abs(power - target_power) <= 0.1:  # Within 10% of target
            return mid
        elif power < target_power:
            low = mid + 1
        else:
            best_estimate = mid
            high = mid - 1
    
    return best_estimate if best_estimate is not None else max_mice

def generate_cache_key(baseline_prop, relative_effects, cluster_sizes, median_icc, target_power, max_mice, n_sims):
    """
    Generate a unique cache key based on analysis parameters
    """
    # Create a string representation of all parameters
    params_str = f"{baseline_prop}_{relative_effects.tostring()}_{cluster_sizes}_{median_icc}_{target_power}_{max_mice}_{n_sims}"
    
    # Create MD5 hash for shorter filename
    cache_key = hashlib.md5(params_str.encode()).hexdigest()
    
    return cache_key

def save_results_to_pickle(results_df, cache_key, cache_dir="power_analysis_cache"):
    """
    Save results DataFrame to pickle file
    """
    # Create cache directory if it doesn't exist
    Path(cache_dir).mkdir(exist_ok=True)
    
    # Create filename
    filename = f"power_analysis_{cache_key}.pkl"
    filepath = Path(cache_dir) / filename
    
    # Save to pickle
    with open(filepath, 'wb') as f:
        pickle.dump(results_df, f)
    
    print(f"Results saved to: {filepath}")
    return filepath

def load_results_from_pickle(cache_key, cache_dir="power_analysis_cache"):
    """
    Load results DataFrame from pickle file if it exists
    """
    filename = f"power_analysis_{cache_key}.pkl"
    filepath = Path(cache_dir) / filename
    
    if filepath.exists():
        try:
            with open(filepath, 'rb') as f:
                results_df = pickle.load(f)
            print(f"Loaded cached results from: {filepath}")
            return results_df
        except Exception as e:
            print(f"Error loading cached results: {e}")
            return None
    else:
        return None

def run_cached_power_analysis(baseline_prop=0.25, cluster_sizes=[50, 100, 2500], 
                             median_icc=0.015, target_power=0.8, max_mice=200,
                             effect_range=(0.05, 0.81), effect_step=0.05,
                             n_sims=100, cache_dir="power_analysis_cache", force_rerun=False):
    """
    Run power analysis with caching - loads from pickle if exists, otherwise runs simulation
    
    Parameters:
    -----------
    baseline_prop : float
        Baseline disease rate
    cluster_sizes : list
        Different cell sampling strategies
    median_icc : float
        Intracluster correlation coefficient
    target_power : float
        Target statistical power
    max_mice : int
        Maximum mice to test
    effect_range : tuple
        (min_effect, max_effect) as proportions
    n_sims : int
        Number of simulations per power calculation (default 100)
    effect_step : float
        Step size for effect sizes
    n_sims : int
        Number of simulations per power calculation (default 100)
    cache_dir : str
        Directory to store cache files
    force_rerun : bool
        If True, ignore cache and rerun analysis
        
    Returns:
    --------
    pandas.DataFrame : Results with required mice per group
    """
    # Create effect size range
    relative_effects = np.arange(effect_range[0], effect_range[1], effect_step)
    
    # Generate cache key
    cache_key = generate_cache_key(baseline_prop, relative_effects, cluster_sizes, 
                                  median_icc, target_power, max_mice, n_sims)
    
    print(f"Power Analysis Parameters:")
    print(f"=" * 40)
    print(f"Baseline disease rate: {baseline_prop*100}%")
    print(f"Target power: {target_power*100}%")
    print(f"ICC: {median_icc:.4f}")
    print(f"Max mice: {max_mice}")
    print(f"Simulations per calculation: {n_sims}")
    print(f"Effect sizes: {len(relative_effects)} from {effect_range[0]*100}% to {(effect_range[1]-effect_step)*100}%")
    print(f"Cluster sizes: {cluster_sizes}")
    print(f"Cache key: {cache_key}")
    print()
    
    # Try to load cached results
    if not force_rerun:
        cached_results = load_results_from_pickle(cache_key, cache_dir)
        if cached_results is not None:
            print("Using cached results - no simulation needed!")
            return cached_results
    
    # Run analysis if no cache or forced rerun
    print("No cached results found - running simulation...")
    print(f"Testing {len(relative_effects)} effect sizes Ã— {len(cluster_sizes)} cell counts")
    
    # Set random seed for reproducibility
    np.random.seed(42)
    
    # Run analysis
    results = []
    total_combinations = len(relative_effects) * len(cluster_sizes)
    
    with tqdm(total=total_combinations, desc="Finding required mice per effect size") as pbar:
        for cluster_size in cluster_sizes:
            for relative_effect in relative_effects:
                required_mice = find_required_mice_for_effect_size(
                    baseline_prop=baseline_prop,
                    relative_effect=relative_effect,
                    cluster_size=cluster_size,
                    median_icc=median_icc,
                    target_power=target_power,
                    max_mice=max_mice,
                    n_sims=n_sims
                )
                
                results.append({
                    'relative_effect_pct': relative_effect * 100,
                    'cluster_size': cluster_size,
                    'required_mice_per_group': required_mice,
                    'baseline_prop': baseline_prop,
                    'median_icc': median_icc,
                    'target_power': target_power,
                    'max_mice': max_mice,
                    'n_sims': n_sims
                })
                
                pbar.update(1)
    
    # Convert to DataFrame
    results_df = pd.DataFrame(results)
    
    # Save results to cache
    save_results_to_pickle(results_df, cache_key, cache_dir)
    
    return results_df

def create_simulation_plot(results_df, figsize=(9, 3)):
    """
    Create plot from simulation results
    """
    baseline_prop = results_df['baseline_prop'].iloc[0]
    target_power = results_df['target_power'].iloc[0]
    median_icc = results_df['median_icc'].iloc[0]
    max_mice = results_df['max_mice'].iloc[0]
    n_sims = results_df['n_sims'].iloc[0]
    
    plt.figure(figsize=figsize)
    
    # Get unique cluster sizes
    cluster_sizes = sorted(results_df['cluster_size'].unique())
    colors = sns.color_palette("deep", n_colors=len(cluster_sizes))
    
    # Plot lines for each cell count
    for i, cluster_size in enumerate(cluster_sizes):
        cluster_data = results_df[results_df['cluster_size'] == cluster_size]
        
        plt.plot(cluster_data['relative_effect_pct'], cluster_data['required_mice_per_group'], 
                marker='o', linewidth=2, markersize=4, color=colors[i],
                label=f'{cluster_size} cells/mouse')
    
    plt.xlabel('Effect Size (% Reduction from Baseline)')
    plt.ylabel('Mice Required per Group')
    plt.title(f'Simulation-Based Sample Size Requirements\n({baseline_prop*100}% Baseline, {target_power*100}% Power, ICC = {median_icc:.4f}, {n_sims} sims)')
    plt.grid(True, alpha=0.3)
    plt.legend(title='Sampling Strategy')
    
    # Set axis limits
    plt.xlim(results_df['relative_effect_pct'].min(), results_df['relative_effect_pct'].max())
    plt.ylim(0, max_mice)
    
    # Add annotation
    plt.text(0.02, 0.98, f'Target: {target_power*100}% Power\nBaseline: {baseline_prop*100}% diseased', 
            transform=plt.gca().transAxes, fontsize=9, verticalalignment='top',
            bbox=dict(boxstyle='round', facecolor='lightgray', alpha=0.8))
    
    # Remove top and right spines
    plt.gca().spines['top'].set_visible(False)
    plt.gca().spines['right'].set_visible(False)
    
    plt.tight_layout()
    plt.show()

def print_simulation_insights(results_df):
    """
    Print key insights from simulation results
    """
    baseline_prop = results_df['baseline_prop'].iloc[0]
    target_power = results_df['target_power'].iloc[0]
    max_mice = results_df['max_mice'].iloc[0]
    
    print(f"\nSimulation Results for {target_power*100}% Power:")
    print("=" * 50)
    
    cluster_sizes = sorted(results_df['cluster_size'].unique())
    
    for cluster_size in cluster_sizes:
        cluster_data = results_df[results_df['cluster_size'] == cluster_size]
        print(f"\n{cluster_size} cells/mouse:")
        for effect in [10, 15, 20, 25, 30, 50]:
            effect_data = cluster_data[cluster_data['relative_effect_pct'] == effect]
            if len(effect_data) > 0:
                mice_needed = effect_data['required_mice_per_group'].iloc[0]
                if mice_needed == max_mice:
                    print(f"  {effect}% effect: >{max_mice} mice per group (underpowered)")
                else:
                    print(f"  {effect}% effect: {mice_needed} mice per group")
    
    # Calculate efficiency gains
    if len(cluster_sizes) >= 2:
        print(f"\nEfficiency comparison (mice saved by using {max(cluster_sizes)} vs {min(cluster_sizes)} cells):")
        for effect in [10, 15, 20, 25, 30, 50]:
            min_data = results_df[(results_df['cluster_size'] == min(cluster_sizes)) & 
                                (results_df['relative_effect_pct'] == effect)]
            max_data = results_df[(results_df['cluster_size'] == max(cluster_sizes)) & 
                                (results_df['relative_effect_pct'] == effect)]
            
            if len(min_data) > 0 and len(max_data) > 0:
                mice_min = min_data['required_mice_per_group'].iloc[0]
                mice_max = max_data['required_mice_per_group'].iloc[0]
                
                if mice_min == max_mice:
                    print(f"  {effect}% effect: {min(cluster_sizes)} cells requires >{max_mice} mice (vs {mice_max} for {max(cluster_sizes)} cells)")
                else:
                    savings = mice_min - mice_max
                    print(f"  {effect}% effect: {savings} fewer mice needed ({mice_min} â†’ {mice_max})")

def clear_cache(cache_dir="power_analysis_cache"):
    """
    Clear all cached results
    """
    cache_path = Path(cache_dir)
    if cache_path.exists():
        for file in cache_path.glob("*.pkl"):
            file.unlink()
        print(f"Cleared cache directory: {cache_dir}")
    else:
        print(f"Cache directory does not exist: {cache_dir}")

# Main execution
if __name__ == "__main__":
    # Analysis parameters
    BASELINE_PROP = 0.25
    CLUSTER_SIZES = [50, 100, 2500]
    MEDIAN_ICC = 0.015  # Use your median ICC here
    TARGET_POWER = 0.8
    MAX_MICE = 200
    N_SIMS = 1000  # Easy to change number of simulations
    
    print("Running Cached Simulation Power Analysis")
    print("=" * 50)
    
    # Run analysis (will load from cache if available)
    mice_required_df = run_cached_power_analysis(
        baseline_prop=BASELINE_PROP,
        cluster_sizes=CLUSTER_SIZES,
        median_icc=MEDIAN_ICC,
        target_power=TARGET_POWER,
        max_mice=MAX_MICE,
        n_sims=N_SIMS
    )
    
    print(f"\nAnalysis completed!")
    print(f"Results stored in 'mice_required_df' DataFrame with {len(mice_required_df)} rows")
    
    # Create visualization
    create_simulation_plot(mice_required_df)
    
    # Print insights
    print_simulation_insights(mice_required_df)
    
    print(f"\nTo change simulations: modify N_SIMS (currently {N_SIMS})")
    print(f"To force rerun: set force_rerun=True")
    print(f"To clear cache: call clear_cache()")
    print(f"Cache files stored in: power_analysis_cache/")
```

```python
import pandas as pd
import numpy as np
from scipy import stats
import statsmodels.api as sm
from statsmodels.genmod import families
from statsmodels.genmod.cov_struct import Exchangeable
from tqdm import tqdm
import matplotlib.pyplot as plt
import seaborn as sns
import pickle
import hashlib
import os
from pathlib import Path

def simulate_clustered_binary_data_fixed_mice(baseline_prop, relative_effect, cluster_size, median_icc, n_clusters=5, n_sims=100):
    """
    Calculate power for fixed number of mice (clusters)
    """
    # Calculate treatment proportion from relative effect
    treatment_prop = baseline_prop * (1 - relative_effect)
    
    # Pre-generate all random effects (vectorized)
    all_cluster_effects_control = np.random.normal(0, np.sqrt(median_icc), (n_sims, n_clusters))
    all_cluster_effects_treatment = np.random.normal(0, np.sqrt(median_icc), (n_sims, n_clusters))
    
    # Pre-calculate logits
    baseline_logit = np.log(baseline_prop / (1 - baseline_prop))
    treatment_logit = np.log(treatment_prop / (1 - treatment_prop))
    
    significant_results = 0
    actual_sims_completed = 0
    
    for sim in range(n_sims):
        # Use pre-generated random effects
        cluster_effects_control = all_cluster_effects_control[sim]
        cluster_effects_treatment = all_cluster_effects_treatment[sim]
        
        control_data = []
        treatment_data = []
        cluster_ids = []
        
        # Generate control group data
        for cluster in range(n_clusters):
            cluster_logit = baseline_logit + cluster_effects_control[cluster]
            cluster_prob = 1 / (1 + np.exp(-cluster_logit))
            cluster_prob = np.clip(cluster_prob, 0.001, 0.999)
            
            cluster_outcomes = np.random.binomial(1, cluster_prob, cluster_size)
            control_data.extend(cluster_outcomes)
            cluster_ids.extend([f'control_{cluster}'] * cluster_size)
        
        # Generate treatment group data
        for cluster in range(n_clusters):
            cluster_logit = treatment_logit + cluster_effects_treatment[cluster]
            cluster_prob = 1 / (1 + np.exp(-cluster_logit))
            cluster_prob = np.clip(cluster_prob, 0.001, 0.999)
            
            cluster_outcomes = np.random.binomial(1, cluster_prob, cluster_size)
            treatment_data.extend(cluster_outcomes)
            cluster_ids.extend([f'treatment_{cluster}'] * cluster_size)
        
        # Combine data
        y = np.array(control_data + treatment_data)
        treatment = np.array([0] * len(control_data) + [1] * len(treatment_data))
        clusters = np.array(cluster_ids)
        
        # Fit GEE model
        try:
            X = sm.add_constant(treatment)
            gee_model = sm.GEE(y, X, groups=clusters, 
                              cov_struct=Exchangeable(), 
                              family=families.Binomial())
            gee_result = gee_model.fit()
            
            # Check if treatment effect is significant
            p_value = gee_result.pvalues[1]
            if p_value < 0.05:
                significant_results += 1
                
        except:
            pass
        
        actual_sims_completed += 1
        
        # Simple early stopping for extreme cases
        if sim >= 15:
            current_power = significant_results / actual_sims_completed
            if current_power == 0.0:
                break
            elif current_power == 1.0 and sim >= 20:
                break
    
    # Calculate final power
    power = significant_results / actual_sims_completed
    power = min(power, 1.0)
    return power

def generate_specific_conditions_cache_key(baseline_prop, relative_effects, conditions, median_icc, n_sims):
    """
    Generate a unique cache key based on specific conditions
    """
    # Create a string representation of all parameters
    conditions_str = str(conditions)
    params_str = f"specific_{baseline_prop}_{relative_effects}_{conditions_str}_{median_icc}_{n_sims}"
    
    # Create MD5 hash for shorter filename
    cache_key = hashlib.md5(params_str.encode()).hexdigest()
    
    return cache_key

def save_specific_results_to_pickle(results_df, cache_key, cache_dir="power_analysis_cache"):
    """
    Save specific conditions results DataFrame to pickle file
    """
    # Create cache directory if it doesn't exist
    cache_path = Path(cache_dir)
    try:
        cache_path.mkdir(parents=True, exist_ok=True)
    except OSError:
        # If cache directory fails, use current directory
        cache_path = Path(".")
        print(f"Warning: Could not create cache directory, using current directory")
    
    # Create filename
    filename = f"specific_conditions_{cache_key}.pkl"
    filepath = cache_path / filename
    
    # Save to pickle
    with open(filepath, 'wb') as f:
        pickle.dump(results_df, f)
    
    print(f"Results saved to: {filepath}")
    return filepath

def load_specific_results_from_pickle(cache_key, cache_dir="power_analysis_cache"):
    """
    Load specific conditions results DataFrame from pickle file if it exists
    """
    filename = f"specific_conditions_{cache_key}.pkl"
    filepath = Path(cache_dir) / filename
    
    if filepath.exists():
        try:
            with open(filepath, 'rb') as f:
                results_df = pickle.load(f)
            print(f"Loaded cached results from: {filepath}")
            return results_df
        except Exception as e:
            print(f"Error loading cached results: {e}")
            return None
    else:
        return None

def run_specific_conditions_analysis(baseline_prop=0.25, median_icc=0.015,
                                   n_sims=1000, cache_dir="power_analysis_cache", 
                                   force_rerun=False):
    """
    Run power analysis for specific conditions:
    - 50 cells with 5 mice per group
    - 100 cells with 5 mice per group
    - 2500 cells with 5 mice per group
    - 50 cells with 100 mice per group
    - 100 cells with 100 mice per group
    
    Returns:
    --------
    pandas.DataFrame : Results with power for each condition
    """
    # Define specific conditions
    conditions = [
        {'cells': 50, 'mice': 6},
        {'cells': 100, 'mice': 6},
        {'cells': 2500, 'mice': 6},
        {'cells': 50, 'mice': 30},
        {'cells': 100, 'mice': 30},
    ]
    
    # Define effect sizes to test
    relative_effects = np.arange(0.05, 0.5, 0.025)  # 5% to 80% in 5% steps
    
    # Generate cache key
    cache_key = generate_specific_conditions_cache_key(baseline_prop, relative_effects.tolist(), 
                                                      conditions, median_icc, n_sims)
    
    print(f"Specific Conditions Power Analysis:")
    print(f"=" * 40)
    print(f"Baseline disease rate: {baseline_prop*100}%")
    print(f"ICC: {median_icc:.4f}")
    print(f"Simulations per calculation: {n_sims}")
    print(f"Testing conditions:")
    for cond in conditions:
        print(f"  - {cond['cells']} cells/mouse, {cond['mice']} mice/group")
    print(f"Effect sizes: {len(relative_effects)} from 5% to 80%")
    print(f"Cache key: {cache_key}")
    print()
    
    # Try to load cached results
    if not force_rerun:
        cached_results = load_specific_results_from_pickle(cache_key, cache_dir)
        if cached_results is not None:
            print("Using cached results - no simulation needed!")
            return cached_results
    
    # Run analysis if no cache or forced rerun
    print("No cached results found - running simulation...")
    print(f"Testing {len(conditions)} conditions Ã— {len(relative_effects)} effect sizes")
    
    # Set random seed for reproducibility
    np.random.seed(42)
    
    # Run analysis
    results = []
    total_combinations = len(conditions) * len(relative_effects)
    
    with tqdm(total=total_combinations, desc="Calculating power for specific conditions") as pbar:
        for condition in conditions:
            cells = condition['cells']
            mice = condition['mice']
            
            for relative_effect in relative_effects:
                power = simulate_clustered_binary_data_fixed_mice(
                    baseline_prop=baseline_prop,
                    relative_effect=relative_effect,
                    cluster_size=cells,
                    median_icc=median_icc,
                    n_clusters=mice,
                    n_sims=n_sims
                )
                
                results.append({
                    'cells_per_mouse': cells,
                    'mice_per_group': mice,
                    'total_cells': cells * mice,
                    'relative_effect_pct': relative_effect * 100,
                    'power': power * 100,  # Convert to percentage
                    'condition_label': f'{cells} cells, {mice} mice',
                    'baseline_prop': baseline_prop,
                    'median_icc': median_icc,
                    'n_sims': n_sims
                })
                
                pbar.update(1)
    
    # Convert to DataFrame
    results_df = pd.DataFrame(results)
    
    # Save results to cache
    save_specific_results_to_pickle(results_df, cache_key, cache_dir)
    
    return results_df

def create_specific_conditions_plot(results_df, figsize=(10, 6)):
    """
    Create plot for specific conditions
    """
    baseline_prop = results_df['baseline_prop'].iloc[0]
    median_icc = results_df['median_icc'].iloc[0]
    n_sims = results_df['n_sims'].iloc[0]
    
    plt.figure(figsize=figsize)
    
    # Get unique conditions
    conditions = results_df['condition_label'].unique()
    colors = sns.color_palette("deep", n_colors=len(conditions))
    
    # Plot lines for each condition
    for i, condition in enumerate(conditions):
        condition_data = results_df[results_df['condition_label'] == condition]
        
        plt.plot(condition_data['relative_effect_pct'], condition_data['power'], 
                marker='o', linewidth=2, markersize=6, color=colors[i],
                label=condition)
    
    # Add horizontal line for 80% power
    plt.axhline(y=80, color='black', linestyle=':', alpha=0.5, label='80% power target')
    
    plt.xlabel('Effect Size (% Reduction from Baseline)')
    plt.ylabel('Statistical Power (%)')
    plt.title(f'Statistical Power for Specific Experimental Conditions\n({baseline_prop*100}% baseline, ICC = {median_icc:.4f}, {n_sims} simulations)')
    plt.grid(True, alpha=0.3)
    plt.legend(loc='lower right')
    
    # Set axis limits
    plt.xlim(0, 85)
    plt.ylim(0, 105)
    
    # Remove top and right spines
    plt.gca().spines['top'].set_visible(False)
    plt.gca().spines['right'].set_visible(False)
    
    plt.tight_layout()
    plt.show()

def print_specific_conditions_insights(results_df):
    """
    Print key insights from specific conditions analysis
    """
    print(f"\nPower Analysis Results for Specific Conditions:")
    print("=" * 60)
    
    # Key effect sizes to highlight
    key_effects = [10, 20, 30, 50]
    
    conditions = results_df['condition_label'].unique()
    
    for effect in key_effects:
        print(f"\n{effect}% effect size:")
        effect_data = results_df[results_df['relative_effect_pct'] == effect]
        
        for condition in conditions:
            cond_data = effect_data[effect_data['condition_label'] == condition]
            if len(cond_data) > 0:
                power = cond_data['power'].iloc[0]
                print(f"  {condition}: {power:.1f}% power")
    
    # Find minimum effect size detectable at 80% power for each condition
    print(f"\nMinimum detectable effect at 80% power:")
    for condition in conditions:
        cond_data = results_df[results_df['condition_label'] == condition]
        # Find first effect size where power >= 80%
        above_80 = cond_data[cond_data['power'] >= 80]
        if len(above_80) > 0:
            min_effect = above_80['relative_effect_pct'].min()
            print(f"  {condition}: {min_effect:.0f}% effect")
        else:
            print(f"  {condition}: >80% effect (underpowered)")

# Main execution
if __name__ == "__main__":
    # Analysis parameters
    BASELINE_PROP = 0.25
    MEDIAN_ICC = 0.015
    N_SIMS = 50  # Number of simulations per power calculation
    
    print("Running Specific Conditions Power Analysis")
    print("=" * 50)
    
    # Run analysis (will load from cache if available)
    specific_conditions_df = run_specific_conditions_analysis(
        baseline_prop=BASELINE_PROP,
        median_icc=MEDIAN_ICC,
        n_sims=N_SIMS
    )
    
    print(f"\nAnalysis completed!")
    print(f"Results stored in 'specific_conditions_df' DataFrame with {len(specific_conditions_df)} rows")
    
    # Create visualization
    create_specific_conditions_plot(specific_conditions_df)
    
    # Print insights
    print_specific_conditions_insights(specific_conditions_df)
    
    print(f"\nDataFrame 'specific_conditions_df' contains columns:")
    print(specific_conditions_df.columns.tolist())
    print(f"\nYou can now use 'specific_conditions_df' for custom plotting in subsequent cells")
    print(f"To force rerun: set force_rerun=True")
    print(f"Cache files stored in: power_analysis_cache/")
```

```python
import pandas as pd
import numpy as np
from scipy import stats
import statsmodels.api as sm
from statsmodels.genmod import families
from statsmodels.genmod.cov_struct import Exchangeable
from tqdm import tqdm
import matplotlib.pyplot as plt
import seaborn as sns
import pickle
import hashlib
from pathlib import Path

def simulate_power_unified(baseline_prop, relative_effect, cluster_size, median_icc, 
                           n_clusters=5, n_sims=100, analysis_type='clustered'):
    """
    Calculate power for fixed number of mice (clusters) using either:
    - 'clustered': GEE analysis accounting for within-mouse correlation
    - 'summary': t-test on mouse-level percentages
    
    Note: n_clusters is the number of mice PER GROUP (control or treatment)
    """
    treatment_prop = baseline_prop * (1 - relative_effect)
    
    all_cluster_effects_control = np.random.normal(0, np.sqrt(median_icc), (n_sims, n_clusters))
    all_cluster_effects_treatment = np.random.normal(0, np.sqrt(median_icc), (n_sims, n_clusters))
    
    baseline_logit = np.log(baseline_prop / (1 - baseline_prop))
    treatment_logit = np.log(treatment_prop / (1 - treatment_prop))
    
    significant_results = 0
    actual_sims_completed = 0
    
    for sim in range(n_sims):
        cluster_effects_control = all_cluster_effects_control[sim]
        cluster_effects_treatment = all_cluster_effects_treatment[sim]
        
        if analysis_type == 'clustered':
            control_data = []
            treatment_data = []
            cluster_ids = []
            
            for cluster in range(n_clusters):
                cluster_logit = baseline_logit + cluster_effects_control[cluster]
                cluster_prob = 1 / (1 + np.exp(-cluster_logit))
                cluster_prob = np.clip(cluster_prob, 0.001, 0.999)
                
                cluster_outcomes = np.random.binomial(1, cluster_prob, cluster_size)
                control_data.extend(cluster_outcomes)
                cluster_ids.extend([f'control_{cluster}'] * cluster_size)
            
            for cluster in range(n_clusters):
                cluster_logit = treatment_logit + cluster_effects_treatment[cluster]
                cluster_prob = 1 / (1 + np.exp(-cluster_logit))
                cluster_prob = np.clip(cluster_prob, 0.001, 0.999)
                
                cluster_outcomes = np.random.binomial(1, cluster_prob, cluster_size)
                treatment_data.extend(cluster_outcomes)
                cluster_ids.extend([f'treatment_{cluster}'] * cluster_size)
            
            y = np.array(control_data + treatment_data)
            treatment = np.array([0] * len(control_data) + [1] * len(treatment_data))
            clusters = np.array(cluster_ids)
            
            try:
                X = sm.add_constant(treatment)
                gee_model = sm.GEE(y, X, groups=clusters, 
                                   cov_struct=Exchangeable(), 
                                   family=families.Binomial())
                gee_result = gee_model.fit()
                
                p_value = gee_result.pvalues[1]
                if p_value < 0.05:
                    significant_results += 1
            except:
                pass
                
        else: # analysis_type == 'summary'
            control_percentages = []
            treatment_percentages = []
            
            for cluster in range(n_clusters):
                cluster_logit = baseline_logit + cluster_effects_control[cluster]
                cluster_prob = 1 / (1 + np.exp(-cluster_logit))
                cluster_prob = np.clip(cluster_prob, 0.001, 0.999)
                
                cluster_outcomes = np.random.binomial(1, cluster_prob, cluster_size)
                cluster_percentage = np.mean(cluster_outcomes)
                control_percentages.append(cluster_percentage)
            
            for cluster in range(n_clusters):
                cluster_logit = treatment_logit + cluster_effects_treatment[cluster]
                cluster_prob = 1 / (1 + np.exp(-cluster_logit))
                cluster_prob = np.clip(cluster_prob, 0.001, 0.999)
                
                cluster_outcomes = np.random.binomial(1, cluster_prob, cluster_size)
                cluster_percentage = np.mean(cluster_outcomes)
                treatment_percentages.append(cluster_percentage)
            
            try:
                _, p_value = stats.ttest_ind(control_percentages, treatment_percentages)
                if p_value < 0.05:
                    significant_results += 1
            except:
                pass
        
        actual_sims_completed += 1
        
        if sim >= 50:
            if actual_sims_completed > 0:
                current_power = significant_results / actual_sims_completed
                if current_power == 0.0:
                    break
                elif current_power == 1.0 and sim >= 50:
                    break
    
    if actual_sims_completed == 0:
        return 0.0

    power = significant_results / actual_sims_completed
    return min(power, 1.0)

def generate_unified_cache_key(baseline_prop, relative_effects, conditions, median_icc, n_sims):
    conditions_str = str(conditions)
    params_str = f"unified_{baseline_prop}_{relative_effects}_{conditions_str}_{median_icc}_{n_sims}"
    return hashlib.md5(params_str.encode()).hexdigest()

def save_unified_results_to_pickle(results_df, cache_key, cache_dir="power_analysis_cache"):
    cache_path = Path(cache_dir)
    cache_path.mkdir(parents=True, exist_ok=True)
    filepath = cache_path / f"unified_power_{cache_key}.pkl"
    with open(filepath, 'wb') as f:
        pickle.dump(results_df, f)
    print(f"Results saved to: {filepath}")
    return filepath

def load_unified_results_from_pickle(cache_key, cache_dir="power_analysis_cache"):
    filepath = Path(cache_dir) / f"unified_power_{cache_key}.pkl"
    if filepath.exists():
        try:
            with open(filepath, 'rb') as f:
                results_df = pickle.load(f)
            print(f"Loaded cached results from: {filepath}")
            return results_df
        except Exception as e:
            print(f"Error loading cached results: {e}")
            return None
    return None

def run_unified_power_analysis(baseline_prop=0.25, median_icc=0.015,
                               n_sims=1000, cache_dir="power_analysis_cache", 
                               force_rerun=False):
    """
    Run unified power analysis comparing clustered (GEE) and summary (t-test) approaches.
    """
    # Define conditions with analysis type
    conditions = [
        # Clustered (GEE) analysis
        {'cells': 50, 'total_mice': 12, 'analysis': 'clustered'},   # 6 mice/group
        {'cells': 100, 'total_mice': 12, 'analysis': 'clustered'},
        {'cells': 2500, 'total_mice': 12, 'analysis': 'clustered'},
        {'cells': 50, 'total_mice': 48, 'analysis': 'clustered'},   # 24 mice/group (was 30/group)
        {'cells': 100, 'total_mice': 48, 'analysis': 'clustered'},  # 24 mice/group (was 30/group)
        
        # Summary (t-test) analysis
        {'cells': 100, 'total_mice': 12, 'analysis': 'summary'},
    ]
    
    relative_effects = np.arange(0.05, 0.51, 0.025)
    
    cache_key = generate_unified_cache_key(baseline_prop, relative_effects.tolist(), 
                                           conditions, median_icc, n_sims)
    
    print("Unified Power Analysis (Clustered vs Summary):")
    print("=" * 50)
    print(f"Baseline disease rate: {baseline_prop*100}%")
    print(f"ICC: {median_icc:.4f}")
    print(f"Simulations per calculation: {n_sims}")
    
    print("\nClustered (GEE) analysis:")
    for cond in conditions:
        if cond['analysis'] == 'clustered':
            print(f"  - {cond['cells']} cells/mouse, {cond['total_mice']} total mice")
    
    print("\nSummary (t-test) analysis:")
    for cond in conditions:
        if cond['analysis'] == 'summary':
            print(f"  - {cond['cells']} cells/mouse, {cond['total_mice']} total mice")
    
    print(f"\nEffect sizes: {len(relative_effects)} from 5% to 47.5%")
    print(f"Cache key: {cache_key}\n")
    
    if not force_rerun:
        cached_results = load_unified_results_from_pickle(cache_key, cache_dir)
        if cached_results is not None:
            print("Using cached results - no simulation needed!")
            return cached_results
    
    print("No cached results found - running simulation...")
    print(f"Testing {len(conditions)} conditions Ã— {len(relative_effects)} effect sizes")
    
    np.random.seed(42)
    results = []
    total_combinations = len(conditions) * len(relative_effects)
    
    with tqdm(total=total_combinations, desc="Calculating power") as pbar:
        for condition in conditions:
            cells = condition['cells']
            total_mice = condition['total_mice']
            mice_per_group = total_mice // 2
            analysis = condition['analysis']
            
            for relative_effect in relative_effects:
                power = simulate_power_unified(
                    baseline_prop=baseline_prop,
                    relative_effect=relative_effect,
                    cluster_size=cells,
                    median_icc=median_icc,
                    n_clusters=mice_per_group,
                    n_sims=n_sims,
                    analysis_type=analysis
                )
                
                if analysis == 'clustered':
                    label = f'{cells} cells, {total_mice} total mice (GEE)'
                else:
                    label = f'{cells} cells, {total_mice} total mice (summary)'
                
                results.append({
                    'cells_per_mouse': cells,
                    'mice_per_group': mice_per_group,
                    'total_mice': total_mice,
                    'relative_effect_pct': relative_effect * 100,
                    'power': power * 100,
                    'analysis_type': analysis,
                    'condition_label': label,
                    'baseline_prop': baseline_prop,
                    'median_icc': median_icc,
                    'n_sims': n_sims
                })
                pbar.update(1)
                
    results_df = pd.DataFrame(results)
    save_unified_results_to_pickle(results_df, cache_key, cache_dir)
    return results_df

def create_unified_comparison_plot(results_df, figsize=(12, 7)):
    baseline_prop = results_df['baseline_prop'].iloc[0]
    median_icc = results_df['median_icc'].iloc[0]
    n_sims = results_df['n_sims'].iloc[0]
    
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=figsize)
    
    colors_clustered = sns.color_palette("viridis", n_colors=5)
    
    ax1.set_title('GEE vs. t-test (12 total mice)')
    
    clustered_12mice = results_df[(results_df['total_mice'] == 12) & (results_df['analysis_type'] == 'clustered')]
    summary_12mice = results_df[(results_df['total_mice'] == 12) & (results_df['analysis_type'] == 'summary')]
    
    cell_counts = sorted(clustered_12mice['cells_per_mouse'].unique())
    for i, cells in enumerate(cell_counts):
        data_clustered = clustered_12mice[clustered_12mice['cells_per_mouse'] == cells]
        if not data_clustered.empty:
            ax1.plot(data_clustered['relative_effect_pct'], data_clustered['power'], 
                     '-o', linewidth=2, markersize=5, color=colors_clustered[i],
                     label=f'{cells} cells (GEE)')
    
    if not summary_12mice.empty:
        total_mice_label = summary_12mice['total_mice'].iloc[0]
        ax1.plot(summary_12mice['relative_effect_pct'], summary_12mice['power'], 
                 '--s', linewidth=2, markersize=5, color='gray',
                 label=f'{summary_12mice["cells_per_mouse"].iloc[0]} cells (t-test)')

    ax1.axhline(y=80, color='black', linestyle=':', alpha=0.7)
    ax1.set_xlabel('Effect Size (% Reduction)')
    ax1.set_ylabel('Statistical Power (%)')
    ax1.grid(True, alpha=0.3)
    ax1.legend(loc='lower right', fontsize=9)
    ax1.set_xlim(0, 50)
    ax1.set_ylim(0, 105)

    ax2.set_title('Power Across All Conditions')
    unique_conditions = sorted(results_df['condition_label'].unique())
    palette = sns.color_palette("deep", len(unique_conditions))
    
    for i, condition in enumerate(unique_conditions):
        condition_data = results_df[results_df['condition_label'] == condition]
        linestyle = '--' if 'summary' in condition else '-'
        marker = 's' if 'summary' in condition else 'o'
        
        ax2.plot(condition_data['relative_effect_pct'], condition_data['power'], 
                 linestyle=linestyle, marker=marker, linewidth=1.5, markersize=4,
                 color=palette[i], label=condition)

    ax2.axhline(y=80, color='black', linestyle=':', alpha=0.7)
    ax2.set_xlabel('Effect Size (% Reduction)')
    ax2.set_ylabel('')
    ax2.grid(True, alpha=0.3)
    ax2.legend(bbox_to_anchor=(1.04, 1), loc='upper left', fontsize=8)
    ax2.set_xlim(0, 50)
    ax2.set_ylim(0, 105)
    
    for ax in [ax1, ax2]:
        ax.spines['top'].set_visible(False)
        ax.spines['right'].set_visible(False)
        
    plt.suptitle(f'Power Analysis: Cell-Level (GEE) vs. Mouse-Level (t-test) Analysis\n'
                 f'({baseline_prop*100}% baseline, ICC = {median_icc:.3f}, {n_sims} simulations)',
                 fontsize=14)
    plt.tight_layout(rect=[0, 0, 0.9, 0.96])
    plt.show()

def print_unified_insights(results_df):
    print("\nPower Analysis Insights:")
    print("=" * 60)
    
    key_effects = [20, 30, 40]
    
    print("\nPower comparison at 12 total mice:")
    for effect in key_effects:
        print(f"\nFor a {effect}% effect size:")
        
        effect_data = results_df[(results_df['relative_effect_pct'] > effect-2) & 
                                 (results_df['relative_effect_pct'] < effect+2) & 
                                 (results_df['total_mice'] == 12)]
        
        for cells in sorted(effect_data['cells_per_mouse'].unique()):
            clustered = effect_data[(effect_data['cells_per_mouse'] == cells) & (effect_data['analysis_type'] == 'clustered')]
            summary = effect_data[(effect_data['cells_per_mouse'] == cells) & (effect_data['analysis_type'] == 'summary')]
            
            if not clustered.empty:
                power_clustered = clustered['power'].iloc[0]
                if not summary.empty:
                    power_summary = summary['power'].iloc[0]
                    diff = power_clustered - power_summary
                    print(f"  {cells:<4} cells: {power_clustered:5.1f}% (GEE) vs {power_summary:5.1f}% (t-test) [+{diff:.1f}% gain]")
                else:
                    print(f"  {cells:<4} cells: {power_clustered:5.1f}% (GEE)")

    print("\nMinimum detectable effect at 80% power:")
    unique_labels = sorted(results_df['condition_label'].unique())
    for label in unique_labels:
        data = results_df[results_df['condition_label'] == label]
        above_80 = data[data['power'] >= 80]
        
        if not above_80.empty:
            min_effect = above_80['relative_effect_pct'].min()
            print(f"  {label:<40}: {min_effect:.1f}% effect")
        else:
            print(f"  {label:<40}: >47.5% effect (underpowered)")

if __name__ == "__main__":
    BASELINE_PROP = 0.25
    MEDIAN_ICC = 0.015
    N_SIMS = 1000
    
    unified_results_df = run_unified_power_analysis(
        baseline_prop=BASELINE_PROP,
        median_icc=MEDIAN_ICC,
        n_sims=N_SIMS,
        force_rerun=False 
    )
    
    print(f"\nAnalysis completed! Results stored in DataFrame with {len(unified_results_df)} rows.")
    
    create_unified_comparison_plot(unified_results_df)
    print_unified_insights(unified_results_df)
```

```python
import pandas as pd
import numpy as np
from scipy import stats
import statsmodels.api as sm
from statsmodels.genmod import families
from statsmodels.genmod.cov_struct import Exchangeable
from tqdm import tqdm
import matplotlib.pyplot as plt
import seaborn as sns
import pickle
import hashlib
import os
from pathlib import Path

def simulate_clustered_binary_data_fixed_mice(baseline_prop, relative_effect, cluster_size, median_icc, n_clusters=5, n_sims=250):
    """
    Calculate power for fixed number of mice (clusters)
    """
    # Calculate treatment proportion from relative effect
    treatment_prop = baseline_prop * (1 - relative_effect)
    
    # Pre-generate all random effects (vectorized)
    all_cluster_effects_control = np.random.normal(0, np.sqrt(median_icc), (n_sims, n_clusters))
    all_cluster_effects_treatment = np.random.normal(0, np.sqrt(median_icc), (n_sims, n_clusters))
    
    # Pre-calculate logits
    baseline_logit = np.log(baseline_prop / (1 - baseline_prop))
    treatment_logit = np.log(treatment_prop / (1 - treatment_prop))
    
    significant_results = 0
    actual_sims_completed = 0
    
    for sim in range(n_sims):
        # Use pre-generated random effects
        cluster_effects_control = all_cluster_effects_control[sim]
        cluster_effects_treatment = all_cluster_effects_treatment[sim]
        
        control_data = []
        treatment_data = []
        cluster_ids = []
        
        # Generate control group data
        for cluster in range(n_clusters):
            cluster_logit = baseline_logit + cluster_effects_control[cluster]
            cluster_prob = 1 / (1 + np.exp(-cluster_logit))
            cluster_prob = np.clip(cluster_prob, 0.001, 0.999)
            
            cluster_outcomes = np.random.binomial(1, cluster_prob, cluster_size)
            control_data.extend(cluster_outcomes)
            cluster_ids.extend([f'control_{cluster}'] * cluster_size)
        
        # Generate treatment group data
        for cluster in range(n_clusters):
            cluster_logit = treatment_logit + cluster_effects_treatment[cluster]
            cluster_prob = 1 / (1 + np.exp(-cluster_logit))
            cluster_prob = np.clip(cluster_prob, 0.001, 0.999)
            
            cluster_outcomes = np.random.binomial(1, cluster_prob, cluster_size)
            treatment_data.extend(cluster_outcomes)
            cluster_ids.extend([f'treatment_{cluster}'] * cluster_size)
        
        # Combine data
        y = np.array(control_data + treatment_data)
        treatment = np.array([0] * len(control_data) + [1] * len(treatment_data))
        clusters = np.array(cluster_ids)
        
        # Fit GEE model
        try:
            X = sm.add_constant(treatment)
            gee_model = sm.GEE(y, X, groups=clusters, 
                              cov_struct=Exchangeable(), 
                              family=families.Binomial())
            gee_result = gee_model.fit()
            
            # Check if treatment effect is significant
            p_value = gee_result.pvalues[1]
            if p_value < 0.05:
                significant_results += 1
                
        except:
            pass
        
        actual_sims_completed += 1
        
        # Simple early stopping for extreme cases
        if sim >= 15:
            current_power = significant_results / actual_sims_completed
            if current_power == 0.0:
                break
            elif current_power == 1.0 and sim >= 20:
                break
    
    # Calculate final power
    power = significant_results / actual_sims_completed
    power = min(power, 1.0)
    return power

def find_required_mice_for_effect_size(baseline_prop, relative_effect, cluster_size, median_icc, target_power=0.8, max_mice=200):
    """
    Find minimum number of mice needed for given effect size using binary search
    """
    if relative_effect == 0:  # No effect, infinite mice needed
        return max_mice
    
    low = 2
    high = max_mice
    best_estimate = None
    
    # Binary search
    while low <= high:
        mid = (low + high) // 2
        
        power = simulate_clustered_binary_data_fixed_mice(
            baseline_prop=baseline_prop,
            relative_effect=relative_effect,
            cluster_size=cluster_size,
            median_icc=median_icc,
            n_clusters=mid,
            n_sims=100
        )
        
        if abs(power - target_power) <= 0.1:  # Within 10% of target
            return mid
        elif power < target_power:
            low = mid + 1
        else:
            best_estimate = mid
            high = mid - 1
    
    return best_estimate if best_estimate is not None else max_mice

def generate_cache_key(baseline_prop, relative_effects, cluster_sizes, median_icc, target_power, max_mice):
    """
    Generate a unique cache key based on analysis parameters
    """
    # Create a string representation of all parameters
    params_str = f"{baseline_prop}_{relative_effects.tostring()}_{cluster_sizes}_{median_icc}_{target_power}_{max_mice}"
    
    # Create MD5 hash for shorter filename
    cache_key = hashlib.md5(params_str.encode()).hexdigest()
    
    return cache_key

def save_results_to_pickle(results_df, cache_key, cache_dir="power_analysis_cache"):
    """
    Save results DataFrame to pickle file
    """
    # Create cache directory if it doesn't exist
    Path(cache_dir).mkdir(exist_ok=True)
    
    # Create filename
    filename = f"power_analysis_{cache_key}.pkl"
    filepath = Path(cache_dir) / filename
    
    # Save to pickle
    with open(filepath, 'wb') as f:
        pickle.dump(results_df, f)
    
    print(f"Results saved to: {filepath}")
    return filepath

def load_results_from_pickle(cache_key, cache_dir="power_analysis_cache"):
    """
    Load results DataFrame from pickle file if it exists
    """
    filename = f"power_analysis_{cache_key}.pkl"
    filepath = Path(cache_dir) / filename
    
    if filepath.exists():
        try:
            with open(filepath, 'rb') as f:
                results_df = pickle.load(f)
            print(f"Loaded cached results from: {filepath}")
            return results_df
        except Exception as e:
            print(f"Error loading cached results: {e}")
            return None
    else:
        return None

def run_cached_power_analysis(baseline_prop=0.25, cluster_sizes=[50, 100, 2500], 
                             median_icc=0.015, target_power=0.8, max_mice=200,
                             effect_range=(0.05, 0.81), effect_step=0.05,
                             cache_dir="power_analysis_cache", force_rerun=False):
    """
    Run power analysis with caching - loads from pickle if exists, otherwise runs simulation
    
    Parameters:
    -----------
    baseline_prop : float
        Baseline disease rate
    cluster_sizes : list
        Different cell sampling strategies
    median_icc : float
        Intracluster correlation coefficient
    target_power : float
        Target statistical power
    max_mice : int
        Maximum mice to test
    effect_range : tuple
        (min_effect, max_effect) as proportions
    effect_step : float
        Step size for effect sizes
    cache_dir : str
        Directory to store cache files
    force_rerun : bool
        If True, ignore cache and rerun analysis
        
    Returns:
    --------
    pandas.DataFrame : Results with required mice per group
    """
    # Create effect size range
    relative_effects = np.arange(effect_range[0], effect_range[1], effect_step)
    
    # Generate cache key
    cache_key = generate_cache_key(baseline_prop, relative_effects, cluster_sizes, 
                                  median_icc, target_power, max_mice)
    
    print(f"Power Analysis Parameters:")
    print(f"=" * 40)
    print(f"Baseline disease rate: {baseline_prop*100}%")
    print(f"Target power: {target_power*100}%")
    print(f"ICC: {median_icc:.4f}")
    print(f"Max mice: {max_mice}")
    print(f"Effect sizes: {len(relative_effects)} from {effect_range[0]*100}% to {(effect_range[1]-effect_step)*100}%")
    print(f"Cluster sizes: {cluster_sizes}")
    print(f"Cache key: {cache_key}")
    print()
    
    # Try to load cached results
    if not force_rerun:
        cached_results = load_results_from_pickle(cache_key, cache_dir)
        if cached_results is not None:
            print("Using cached results - no simulation needed!")
            return cached_results
    
    # Run analysis if no cache or forced rerun
    print("No cached results found - running simulation...")
    print(f"Testing {len(relative_effects)} effect sizes Ã— {len(cluster_sizes)} cell counts")
    
    # Set random seed for reproducibility
    np.random.seed(42)
    
    # Run analysis
    results = []
    total_combinations = len(relative_effects) * len(cluster_sizes)
    
    with tqdm(total=total_combinations, desc="Finding required mice per effect size") as pbar:
        for cluster_size in cluster_sizes:
            for relative_effect in relative_effects:
                required_mice = find_required_mice_for_effect_size(
                    baseline_prop=baseline_prop,
                    relative_effect=relative_effect,
                    cluster_size=cluster_size,
                    median_icc=median_icc,
                    target_power=target_power,
                    max_mice=max_mice
                )
                
                results.append({
                    'relative_effect_pct': relative_effect * 100,
                    'cluster_size': cluster_size,
                    'required_mice_per_group': required_mice,
                    'baseline_prop': baseline_prop,
                    'median_icc': median_icc,
                    'target_power': target_power,
                    'max_mice': max_mice
                })
                
                pbar.update(1)
    
    # Convert to DataFrame
    results_df = pd.DataFrame(results)
    
    # Save results to cache
    save_results_to_pickle(results_df, cache_key, cache_dir)
    
    return results_df

def create_simulation_plot(results_df, figsize=(9, 3)):
    """
    Create plot from simulation results
    """
    baseline_prop = results_df['baseline_prop'].iloc[0]
    target_power = results_df['target_power'].iloc[0]
    median_icc = results_df['median_icc'].iloc[0]
    max_mice = results_df['max_mice'].iloc[0]
    
    plt.figure(figsize=figsize)
    
    # Get unique cluster sizes
    cluster_sizes = sorted(results_df['cluster_size'].unique())
    colors = sns.color_palette("deep", n_colors=len(cluster_sizes))
    
    # Plot lines for each cell count
    for i, cluster_size in enumerate(cluster_sizes):
        cluster_data = results_df[results_df['cluster_size'] == cluster_size]
        
        plt.plot(cluster_data['relative_effect_pct'], cluster_data['required_mice_per_group'], 
                marker='o', linewidth=2, markersize=4, color=colors[i],
                label=f'{cluster_size} cells/mouse')
    
    plt.xlabel('Effect Size (% Reduction from Baseline)')
    plt.ylabel('Mice Required per Group')
    plt.title(f'Simulation-Based Sample Size Requirements\n({baseline_prop*100}% Baseline, {target_power*100}% Power, ICC = {median_icc:.4f})')
    plt.grid(True, alpha=0.3)
    plt.legend(title='Sampling Strategy')
    
    # Set axis limits
    plt.xlim(results_df['relative_effect_pct'].min(), results_df['relative_effect_pct'].max())
    plt.ylim(0, max_mice)
    
    # Add annotation
    plt.text(0.02, 0.98, f'Target: {target_power*100}% Power\nBaseline: {baseline_prop*100}% diseased', 
            transform=plt.gca().transAxes, fontsize=9, verticalalignment='top',
            bbox=dict(boxstyle='round', facecolor='lightgray', alpha=0.8))
    
    # Remove top and right spines
    plt.gca().spines['top'].set_visible(False)
    plt.gca().spines['right'].set_visible(False)
    
    plt.tight_layout()
    plt.show()

def print_simulation_insights(results_df):
    """
    Print key insights from simulation results
    """
    baseline_prop = results_df['baseline_prop'].iloc[0]
    target_power = results_df['target_power'].iloc[0]
    max_mice = results_df['max_mice'].iloc[0]
    
    print(f"\nSimulation Results for {target_power*100}% Power:")
    print("=" * 50)
    
    cluster_sizes = sorted(results_df['cluster_size'].unique())
    
    for cluster_size in cluster_sizes:
        cluster_data = results_df[results_df['cluster_size'] == cluster_size]
        print(f"\n{cluster_size} cells/mouse:")
        for effect in [10, 20, 30, 50]:
            effect_data = cluster_data[cluster_data['relative_effect_pct'] == effect]
            if len(effect_data) > 0:
                mice_needed = effect_data['required_mice_per_group'].iloc[0]
                if mice_needed == max_mice:
                    print(f"  {effect}% effect: >{max_mice} mice per group (underpowered)")
                else:
                    print(f"  {effect}% effect: {mice_needed} mice per group")
    
    # Calculate efficiency gains
    if len(cluster_sizes) >= 2:
        print(f"\nEfficiency comparison (mice saved by using {max(cluster_sizes)} vs {min(cluster_sizes)} cells):")
        for effect in [10, 20, 30, 50]:
            min_data = results_df[(results_df['cluster_size'] == min(cluster_sizes)) & 
                                (results_df['relative_effect_pct'] == effect)]
            max_data = results_df[(results_df['cluster_size'] == max(cluster_sizes)) & 
                                (results_df['relative_effect_pct'] == effect)]
            
            if len(min_data) > 0 and len(max_data) > 0:
                mice_min = min_data['required_mice_per_group'].iloc[0]
                mice_max = max_data['required_mice_per_group'].iloc[0]
                
                if mice_min == max_mice:
                    print(f"  {effect}% effect: {min(cluster_sizes)} cells requires >{max_mice} mice (vs {mice_max} for {max(cluster_sizes)} cells)")
                else:
                    savings = mice_min - mice_max
                    print(f"  {effect}% effect: {savings} fewer mice needed ({mice_min} â†’ {mice_max})")

def clear_cache(cache_dir="power_analysis_cache"):
    """
    Clear all cached results
    """
    cache_path = Path(cache_dir)
    if cache_path.exists():
        for file in cache_path.glob("*.pkl"):
            file.unlink()
        print(f"Cleared cache directory: {cache_dir}")
    else:
        print(f"Cache directory does not exist: {cache_dir}")

# Main execution
if __name__ == "__main__":
    # Analysis parameters
    BASELINE_PROP = 0.25
    CLUSTER_SIZES = [50, 100, 2500]
    MEDIAN_ICC = 0.015  # Use your median ICC here
    TARGET_POWER = 0.8
    MAX_MICE = 200
    
    print("Running Cached Simulation Power Analysis")
    print("=" * 50)
    
    # Run analysis (will load from cache if available)
    mice_required_df = run_cached_power_analysis(
        baseline_prop=BASELINE_PROP,
        cluster_sizes=CLUSTER_SIZES,
        median_icc=MEDIAN_ICC,
        target_power=TARGET_POWER,
        max_mice=MAX_MICE
    )
    
    print(f"\nAnalysis completed!")
    print(f"Results stored in 'mice_required_df' DataFrame with {len(mice_required_df)} rows")
    
    # Create visualization
    create_simulation_plot(mice_required_df)
    
    # Print insights
    print_simulation_insights(mice_required_df)
    
    print(f"\nTo force rerun: set force_rerun=True")
    print(f"To clear cache: call clear_cache()")
    print(f"Cache files stored in: power_analysis_cache/")
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
from tqdm import tqdm

def logit_icc_to_proportion_icc(icc_logit, baseline_prop):
    """
    Convert ICC from logit scale to proportion scale
    
    Parameters:
    -----------
    icc_logit : float
        ICC on the logit scale (from random effects model)
    baseline_prop : float
        Baseline proportion for conversion
        
    Returns:
    --------
    float : ICC on proportion scale
    """
    # Avoid division by zero and extreme values
    icc_logit = np.clip(icc_logit, 0.001, 0.999)
    
    # Variance of logistic distribution
    logistic_var = np.pi**2 / 3
    
    # Between-cluster variance on logit scale
    sigma_b_squared = icc_logit * logistic_var / (1 - icc_logit)
    
    # Convert to proportion scale using delta method
    # Derivative of logistic function at baseline
    derivative = baseline_prop * (1 - baseline_prop)
    
    # Proportion-scale between-cluster variance
    var_between_prop = (derivative**2) * sigma_b_squared
    
    # Total variance on proportion scale  
    var_within_prop = baseline_prop * (1 - baseline_prop)
    var_total_prop = var_between_prop + var_within_prop
    
    # ICC on proportion scale
    icc_proportion = var_between_prop / var_total_prop
    
    return min(icc_proportion, 0.99)  # Cap at reasonable maximum

def calculate_individual_sample_size(baseline_prop, relative_effect, alpha=0.05, power=0.8):
    """
    Calculate sample size for individual-level binary outcome (ignoring clustering)
    """
    # Calculate treatment proportion
    treatment_prop = baseline_prop * (1 - relative_effect)
    
    # Effect size (absolute difference in proportions)
    effect_size = abs(baseline_prop - treatment_prop)
    
    # Handle edge case: no effect
    if effect_size == 0:
        return 999999  # Infinite sample size
    
    # Pooled proportion for variance calculation
    p_pooled = (baseline_prop + treatment_prop) / 2
    
    # Critical values
    z_alpha = stats.norm.ppf(1 - alpha/2)  # Two-sided test
    z_beta = stats.norm.ppf(power)
    
    # Sample size formula for comparing two proportions
    numerator = 2 * p_pooled * (1 - p_pooled) * (z_alpha + z_beta)**2
    denominator = effect_size**2
    
    n_individual = numerator / denominator
    
    return int(np.ceil(n_individual))

def calculate_design_effect_corrected(cluster_size, icc_logit, baseline_prop):
    """
    Calculate design effect using logit-scale ICC conversion
    
    Parameters:
    -----------
    cluster_size : int
        Number of observations per cluster
    icc_logit : float
        ICC on logit scale (from your simulation)
    baseline_prop : float
        Baseline proportion (needed for conversion)
        
    Returns:
    --------
    float : Design effect multiplier
    """
    # Convert logit-scale ICC to proportion-scale ICC
    icc_proportion = logit_icc_to_proportion_icc(icc_logit, baseline_prop)
    
    # Standard design effect formula
    design_effect = 1 + (cluster_size - 1) * icc_proportion
    
    return design_effect

def analytical_mice_required_corrected(baseline_prop, relative_effect, cluster_size, 
                                     icc_logit, alpha=0.05, power=0.8):
    """
    Calculate required mice using corrected ICC conversion
    
    Parameters:
    -----------
    baseline_prop : float
        Baseline disease rate (proportion)
    relative_effect : float
        Relative reduction from baseline
    cluster_size : int
        Number of cells per mouse
    icc_logit : float
        ICC on logit scale (from your random effects simulation)
    alpha : float
        Type I error rate
    power : float
        Statistical power
        
    Returns:
    --------
    int : Number of mice required per group
    """
    # Step 1: Calculate individual-level sample size
    n_individual = calculate_individual_sample_size(baseline_prop, relative_effect, alpha, power)
    
    # Step 2: Apply corrected design effect
    design_effect = calculate_design_effect_corrected(cluster_size, icc_logit, baseline_prop)
    n_clustered_total = n_individual * design_effect
    
    # Step 3: Convert to number of clusters (mice)
    n_mice = n_clustered_total / cluster_size
    
    return int(np.ceil(n_mice))

def run_corrected_analytical_analysis(baseline_prop=0.25, cluster_sizes=[50, 100, 2500],
                                    icc_logit=0.015, target_power=0.8, 
                                    effect_range=(0.05, 0.81), effect_step=0.05):
    """
    Run analytical analysis with logit-scale ICC correction
    
    Parameters:
    -----------
    baseline_prop : float
        Baseline disease rate
    cluster_sizes : list
        Different cell sampling strategies
    icc_logit : float
        ICC on logit scale (from your simulation model)
    target_power : float
        Target statistical power
    effect_range : tuple
        (min_effect, max_effect) as proportions
    effect_step : float
        Step size for effect sizes
        
    Returns:
    --------
    pandas.DataFrame : Results with corrected ICC conversion
    """
    # Create effect size range
    relative_effects = np.arange(effect_range[0], effect_range[1], effect_step)
    
    # Convert logit ICC to proportion ICC for display
    icc_proportion = logit_icc_to_proportion_icc(icc_logit, baseline_prop)
    
    print(f"Corrected Analytical Power Analysis")
    print(f"=" * 40)
    print(f"Baseline disease rate: {baseline_prop*100}%")
    print(f"Target power: {target_power*100}%")
    print(f"ICC (logit scale): {icc_logit:.4f}")
    print(f"ICC (proportion scale): {icc_proportion:.4f}")
    print(f"Testing {len(relative_effects)} effect sizes Ã— {len(cluster_sizes)} sampling strategies")
    print()
    
    # Run analysis
    results = []
    total_combinations = len(relative_effects) * len(cluster_sizes)
    
    with tqdm(total=total_combinations, desc="Calculating corrected sample sizes") as pbar:
        for cluster_size in cluster_sizes:
            # Calculate design effect for this cluster size
            design_effect = calculate_design_effect_corrected(cluster_size, icc_logit, baseline_prop)
            
            for relative_effect in relative_effects:
                required_mice = analytical_mice_required_corrected(
                    baseline_prop=baseline_prop,
                    relative_effect=relative_effect,
                    cluster_size=cluster_size,
                    icc_logit=icc_logit,
                    power=target_power
                )
                
                # Cap at maximum
                required_mice = min(required_mice, 200)
                
                results.append({
                    'relative_effect_pct': relative_effect * 100,
                    'cluster_size': cluster_size,
                    'required_mice_per_group': required_mice,
                    'design_effect': design_effect,
                    'icc_logit': icc_logit,
                    'icc_proportion': icc_proportion,
                    'baseline_prop': baseline_prop,
                    'target_power': target_power
                })
                
                pbar.update(1)
    
    return pd.DataFrame(results)

def create_corrected_plot(results_df, figsize=(9, 3)):
    """
    Create plot with corrected ICC conversion
    """
    baseline_prop = results_df['baseline_prop'].iloc[0]
    target_power = results_df['target_power'].iloc[0]
    icc_logit = results_df['icc_logit'].iloc[0]
    icc_proportion = results_df['icc_proportion'].iloc[0]
    
    plt.figure(figsize=figsize)
    
    # Get unique cluster sizes
    cluster_sizes = sorted(results_df['cluster_size'].unique())
    colors = sns.color_palette("deep", n_colors=len(cluster_sizes))
    
    # Plot lines for each sampling strategy
    for i, cluster_size in enumerate(cluster_sizes):
        cluster_data = results_df[results_df['cluster_size'] == cluster_size]
        design_effect = cluster_data['design_effect'].iloc[0]
        
        plt.plot(cluster_data['relative_effect_pct'], cluster_data['required_mice_per_group'],
                marker='o', linewidth=2, markersize=4, color=colors[i],
                label=f'{cluster_size} cells/mouse (DE={design_effect:.1f})')
    
    plt.xlabel('Effect Size (% Reduction from Baseline)')
    plt.ylabel('Mice Required per Group')
    plt.title(f'Corrected Analytical Sample Size Requirements\n'
              f'({baseline_prop*100}% Baseline, {target_power*100}% Power, '
              f'ICC={icc_logit:.3f} logit â†’ {icc_proportion:.3f} proportion)')
    plt.grid(True, alpha=0.3)
    plt.legend(title='Sampling Strategy', fontsize=9)
    
    # Set limits
    plt.xlim(results_df['relative_effect_pct'].min(), results_df['relative_effect_pct'].max())
    plt.ylim(0, 200)
    
    # Clean up plot
    plt.gca().spines['top'].set_visible(False)
    plt.gca().spines['right'].set_visible(False)
    
    plt.tight_layout()
    plt.show()

def compare_icc_conversions(icc_logit, baseline_props=[0.1, 0.25, 0.5, 0.75]):
    """
    Show how ICC conversion varies with baseline proportion
    """
    print(f"ICC Conversion for Different Baseline Proportions")
    print(f"=" * 50)
    print(f"Logit-scale ICC: {icc_logit:.4f}")
    print()
    
    for baseline_prop in baseline_props:
        icc_prop = logit_icc_to_proportion_icc(icc_logit, baseline_prop)
        print(f"Baseline {baseline_prop*100:2.0f}%: ICC(proportion) = {icc_prop:.4f}")
    
    print(f"\nNote: ICC conversion depends on baseline proportion!")
    print(f"Use the baseline proportion from your specific study.")

# Example usage
if __name__ == "__main__":
    # Analysis parameters - using your simulation's logit-scale ICC
    BASELINE_PROP = 0.25
    CLUSTER_SIZES = [50, 100, 2500]
    ICC_LOGIT = 0.015  # This is from your simulation (logit scale)
    TARGET_POWER = 0.8
    
    print("Running Corrected Analytical Power Analysis...")
    print("(Using logit-scale ICC conversion)")
    
    # Show ICC conversion
    compare_icc_conversions(ICC_LOGIT)
    print()
    
    # Run corrected analysis
    corrected_results = run_corrected_analytical_analysis(
        baseline_prop=BASELINE_PROP,
        cluster_sizes=CLUSTER_SIZES,
        icc_logit=ICC_LOGIT,
        target_power=TARGET_POWER
    )
    
    print(f"\nCorrected analysis completed!")
    print(f"Results stored in DataFrame with {len(corrected_results)} rows")
    
    # Create visualization
    create_corrected_plot(corrected_results)
    
    print(f"\nThis should match your simulation results much better!")
    print(f"The key was converting from logit-scale ICC to proportion-scale ICC.")
```

```python
import pandas as pd
import numpy as np
from scipy import stats
import statsmodels.api as sm
from statsmodels.genmod import families
from statsmodels.genmod.cov_struct import Exchangeable
from tqdm import tqdm
import matplotlib.pyplot as plt
import seaborn as sns

def simulate_clustered_binary_data_vectorized(baseline_prop, relative_effect, cluster_size, 
                                            median_icc, n_clusters=5, n_sims=100):
    """
    Vectorized data generation for clustered binary data with sequential GEE fitting
    
    Parameters:
    -----------
    baseline_prop : float
        Baseline proportion of positive outcomes
    relative_effect : float  
        Relative reduction (e.g., 0.2 = 20% reduction)
    cluster_size : int
        Number of observations per cluster
    median_icc : float
        Intracluster correlation coefficient
    n_clusters : int
        Number of clusters per group
    n_sims : int
        Number of simulations to run
    
    Returns:
    --------
    float : Estimated power
    """
    # Calculate treatment proportion
    treatment_prop = baseline_prop * (1 - relative_effect)
    
    # Pre-calculate logits
    baseline_logit = np.log(baseline_prop / (1 - baseline_prop))
    treatment_logit = np.log(treatment_prop / (1 - treatment_prop))
    
    # Generate ALL random effects at once: shape (n_sims, n_clusters)
    cluster_effects_control = np.random.normal(0, np.sqrt(median_icc), (n_sims, n_clusters))
    cluster_effects_treatment = np.random.normal(0, np.sqrt(median_icc), (n_sims, n_clusters))
    
    # Calculate ALL cluster probabilities at once using broadcasting
    control_logits = baseline_logit + cluster_effects_control  # (n_sims, n_clusters)
    treatment_logits = treatment_logit + cluster_effects_treatment  # (n_sims, n_clusters)
    
    control_probs = 1 / (1 + np.exp(-control_logits))
    treatment_probs = 1 / (1 + np.exp(-treatment_logits))
    
    # Clip probabilities to avoid numerical issues
    control_probs = np.clip(control_probs, 0.001, 0.999)
    treatment_probs = np.clip(treatment_probs, 0.001, 0.999)
    
    # Generate ALL outcomes at once: shape (n_sims, n_clusters, cluster_size)
    control_outcomes = np.random.binomial(1, control_probs[:, :, np.newaxis], 
                                        (n_sims, n_clusters, cluster_size))
    treatment_outcomes = np.random.binomial(1, treatment_probs[:, :, np.newaxis], 
                                          (n_sims, n_clusters, cluster_size))
    
    # Pre-create cluster IDs (same for all simulations)
    total_obs_per_group = n_clusters * cluster_size
    cluster_ids_base = np.repeat(np.arange(n_clusters), cluster_size)
    control_cluster_ids = [f'control_{i}' for i in cluster_ids_base]
    treatment_cluster_ids = [f'treatment_{i}' for i in cluster_ids_base]
    all_cluster_ids = np.array(control_cluster_ids + treatment_cluster_ids)
    
    # Pre-create treatment indicator and design matrix
    treatment_indicator = np.array([0] * total_obs_per_group + [1] * total_obs_per_group)
    X = sm.add_constant(treatment_indicator)
    
    # Sequential GEE fitting for each simulation
    significant_results = 0
    successful_fits = 0
    
    for sim in range(n_sims):
        try:
            # Extract and flatten data for this simulation
            control_data = control_outcomes[sim].flatten()
            treatment_data = treatment_outcomes[sim].flatten()
            
            # Combine data
            y = np.concatenate([control_data, treatment_data])
            
            # Fit GEE model
            gee_model = sm.GEE(y, X, groups=all_cluster_ids,
                             cov_struct=Exchangeable(),
                             family=families.Binomial())
            gee_result = gee_model.fit()
            
            # Check significance
            if gee_result.pvalues[1] < 0.05:
                significant_results += 1
                
        except:
            pass
        
        successful_fits += 1
        
        # Early stopping for extreme cases
        if sim >= 15:
            current_power = significant_results / successful_fits
            if current_power == 0.0:
                break
            elif current_power == 1.0 and sim >= 20:
                break
    
    power = significant_results / successful_fits if successful_fits > 0 else 0.0
    return min(power, 1.0)

def find_required_mice_for_effect_size(baseline_prop, relative_effect, cluster_size, 
                                     median_icc, target_power=0.8, max_mice=100, 
                                     n_sims=100):
    """
    Find minimum number of mice needed for given effect size using binary search
    """
    if relative_effect == 0:
        return max_mice
    
    low = 2
    high = max_mice
    best_estimate = None
    
    # Binary search
    while low <= high:
        mid = (low + high) // 2
        
        power = simulate_clustered_binary_data_vectorized(
            baseline_prop=baseline_prop,
            relative_effect=relative_effect,
            cluster_size=cluster_size,
            median_icc=median_icc,
            n_clusters=mid,
            n_sims=n_sims
        )
        
        if abs(power - target_power) <= 0.1:
            return mid
        elif power < target_power:
            low = mid + 1
        else:
            best_estimate = mid
            high = mid - 1
    
    return best_estimate if best_estimate is not None else max_mice

def run_complete_power_analysis(baseline_prop=0.25, cluster_sizes=[50, 100, 2500], 
                              median_icc=0.05, target_power=0.8, n_sims=100,
                              effect_range=(0.05, 0.81), effect_step=0.05):
    """
    Run complete power analysis across effect sizes and sampling strategies
    
    Parameters:
    -----------
    baseline_prop : float
        Baseline disease rate (proportion)
    cluster_sizes : list
        Different cell sampling strategies (cells per mouse)
    median_icc : float
        Intracluster correlation coefficient
    target_power : float
        Target statistical power (default 0.8 = 80%)
    n_sims : int
        Number of simulations per power calculation
    effect_range : tuple
        (min_effect, max_effect) as proportions
    effect_step : float
        Step size for effect sizes
        
    Returns:
    --------
    pandas.DataFrame : Results with required mice per group for each scenario
    """
    
    # Create effect size range
    relative_effects = np.arange(effect_range[0], effect_range[1], effect_step)
    
    print(f"Running Complete Power Analysis")
    print(f"=" * 50)
    print(f"Baseline disease rate: {baseline_prop*100}%")
    print(f"Target power: {target_power*100}%")
    print(f"ICC: {median_icc:.4f}")
    print(f"Simulations per analysis: {n_sims}")
    print(f"Testing {len(relative_effects)} effect sizes Ã— {len(cluster_sizes)} sampling strategies")
    print()
    
    # Run analysis
    results = []
    total_combinations = len(relative_effects) * len(cluster_sizes)
    
    with tqdm(total=total_combinations, desc="Analyzing power requirements") as pbar:
        for cluster_size in cluster_sizes:
            for relative_effect in relative_effects:
                required_mice = find_required_mice_for_effect_size(
                    baseline_prop=baseline_prop,
                    relative_effect=relative_effect,
                    cluster_size=cluster_size,
                    median_icc=median_icc,
                    target_power=target_power,
                    max_mice=100,
                    n_sims=n_sims
                )
                
                results.append({
                    'relative_effect_pct': relative_effect * 100,
                    'cluster_size': cluster_size,
                    'required_mice_per_group': required_mice,
                    'baseline_prop': baseline_prop,
                    'target_power': target_power,
                    'median_icc': median_icc
                })
                
                pbar.update(1)
    
    return pd.DataFrame(results)

def create_power_analysis_plot(results_df, figsize=(9, 3)):
    """
    Create publication-ready plot of power analysis results
    """
    baseline_prop = results_df['baseline_prop'].iloc[0]
    target_power = results_df['target_power'].iloc[0]
    median_icc = results_df['median_icc'].iloc[0]
    
    plt.figure(figsize=figsize)
    
    # Get unique cluster sizes and assign colors
    cluster_sizes = sorted(results_df['cluster_size'].unique())
    colors = sns.color_palette("deep", n_colors=len(cluster_sizes))
    
    # Plot lines for each cell count
    for i, cluster_size in enumerate(cluster_sizes):
        cluster_data = results_df[results_df['cluster_size'] == cluster_size]
        
        plt.plot(cluster_data['relative_effect_pct'], cluster_data['required_mice_per_group'], 
                marker='o', linewidth=2, markersize=4, color=colors[i],
                label=f'{cluster_size} cells/mouse')
    
    plt.xlabel('Effect Size (% Reduction from Baseline)')
    plt.ylabel('Mice Required per Group')
    plt.title(f'Sample Size Requirements vs Effect Size\n({baseline_prop*100}% Baseline, {target_power*100}% Power, ICC = {median_icc:.4f})')
    plt.grid(True, alpha=0.3)
    plt.legend(title='Sampling Strategy')
    
    # Set axis limits
    plt.xlim(results_df['relative_effect_pct'].min(), results_df['relative_effect_pct'].max())
    plt.ylim(0, 100)
    
    # Remove top and right spines
    plt.gca().spines['top'].set_visible(False)
    plt.gca().spines['right'].set_visible(False)
    
    plt.tight_layout()
    plt.show()

def print_key_insights(results_df):
    """
    Print key insights from the power analysis
    """
    baseline_prop = results_df['baseline_prop'].iloc[0]
    target_power = results_df['target_power'].iloc[0]
    
    print(f"\nKey Insights for {target_power*100}% Power:")
    print("=" * 50)
    
    cluster_sizes = sorted(results_df['cluster_size'].unique())
    
    for cluster_size in cluster_sizes:
        cluster_data = results_df[results_df['cluster_size'] == cluster_size]
        print(f"\n{cluster_size} cells/mouse:")
        
        for effect in [10, 20, 30, 50]:
            effect_data = cluster_data[cluster_data['relative_effect_pct'] == effect]
            if len(effect_data) > 0:
                mice_needed = effect_data['required_mice_per_group'].iloc[0]
                if mice_needed == 100:
                    print(f"  {effect}% effect: >100 mice per group (underpowered)")
                else:
                    print(f"  {effect}% effect: {mice_needed} mice per group")
    
    # Calculate efficiency gains
    if len(cluster_sizes) >= 2:
        print(f"\nEfficiency Comparison (mice saved by using {max(cluster_sizes)} vs {min(cluster_sizes)} cells):")
        print("-" * 70)
        
        for effect in [10, 20, 30, 50]:
            min_size_data = results_df[(results_df['cluster_size'] == min(cluster_sizes)) & 
                                     (results_df['relative_effect_pct'] == effect)]
            max_size_data = results_df[(results_df['cluster_size'] == max(cluster_sizes)) & 
                                     (results_df['relative_effect_pct'] == effect)]
            
            if len(min_size_data) > 0 and len(max_size_data) > 0:
                mice_min = min_size_data['required_mice_per_group'].iloc[0]
                mice_max = max_size_data['required_mice_per_group'].iloc[0]
                
                if mice_min == 100:
                    print(f"  {effect}% effect: {min(cluster_sizes)} cells requires >100 mice (vs {mice_max} for {max(cluster_sizes)} cells)")
                else:
                    savings = mice_min - mice_max
                    print(f"  {effect}% effect: {savings} fewer mice needed ({mice_min} â†’ {mice_max})")

# Example usage and complete analysis
if __name__ == "__main__":
    # Set random seed for reproducibility
    np.random.seed(42)
    
    # Define analysis parameters
    BASELINE_PROP = 0.25  # 25% baseline disease rate
    CLUSTER_SIZES = [50, 100, 2500]  # Different sampling strategies
    MEDIAN_ICC = 0.015  # Example ICC value
    TARGET_POWER = 0.8  # 80% power
    N_SIMS = 10  # Number of simulations (increase for more precision)
    
    print("Starting Complete Clustered Mouse Study Power Analysis")
    print("=" * 60)
    
    # Run the complete analysis
    results = run_complete_power_analysis(
        baseline_prop=BASELINE_PROP,
        cluster_sizes=CLUSTER_SIZES,
        median_icc=MEDIAN_ICC,
        target_power=TARGET_POWER,
        n_sims=N_SIMS,
        effect_range=(0.05, 0.81),
        effect_step=0.05
    )
    
    print("\nAnalysis completed!")
    print(f"Results stored in DataFrame with {len(results)} rows")
    
    # Create visualization
    print("\nCreating visualization...")
    create_power_analysis_plot(results)
    
    # Print insights
    print_key_insights(results)
    
    print(f"\nAnalysis complete! Use 'results' DataFrame for further analysis.")
    print(f"To change number of simulations, modify N_SIMS parameter (currently {N_SIMS})")
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import subprocess
import os
import time
from tqdm import tqdm
import glob
from multiprocessing import Pool

# =============================================================================
# CONFIGURABLE PARAMETERS
# =============================================================================

BASELINE_PROP = 0.25        # Baseline disease rate (25%)
MAX_MICE_TESTED = 100       # Maximum number of mice to test per group
N_SIMULATIONS = 10        # Number of simulations per power calculation
RELATIVE_EFFECTS = np.arange(0.05, 0.81, 0.05)  # 5% to 80% in 5% increments
CLUSTER_SIZES = [50, 100, 2500]                 # Cells per mouse
TARGET_POWER = 0.8          # Target statistical power (80%)
N_CORES = 28                # Number of parallel processes

def get_median_icc(icc_results):
    """Extract median ICC from results"""
    all_icc_values = []
    for dataset in icc_results.keys():
        for treatment in icc_results[dataset].keys():
            if treatment not in ['overall', 'Wildtype']:
                for outcome in icc_results[dataset][treatment].keys():
                    icc_val = icc_results[dataset][treatment][outcome]
                    if icc_val is not None and not np.isnan(icc_val):
                        all_icc_values.append(icc_val)
    return np.median(all_icc_values)

def create_job_list(baseline_prop, relative_effects, cluster_sizes, median_icc, 
                   target_power, max_mice, n_sims):
    """Create list of all command line calls"""
    jobs = []
    job_id = 0
    
    for cluster_size in cluster_sizes:
        for relative_effect in relative_effects:
            cmd = [
                'python', 'power_simulation_single.py',
                str(baseline_prop),
                str(relative_effect),
                str(cluster_size),
                str(median_icc),
                str(target_power),
                str(max_mice),
                str(n_sims),
                f'job_{job_id:03d}'
            ]
            jobs.append(cmd)
            job_id += 1
    
    return jobs

def run_single_subprocess(cmd):
    """Run a single subprocess command"""
    try:
        result = subprocess.run(cmd, capture_output=True, text=True, timeout=600)  # 10 min timeout
        if result.returncode == 0:
            return True, result.stdout.strip()
        else:
            return False, result.stderr.strip()
    except subprocess.TimeoutExpired:
        return False, "Process timed out"
    except Exception as e:
        return False, str(e)

def clean_results_directory():
    """Clean up any existing results"""
    if os.path.exists('power_sim_results'):
        for file in glob.glob('power_sim_results/*.csv'):
            os.remove(file)
        print("Cleaned existing results directory")
    else:
        os.makedirs('power_sim_results', exist_ok=True)
        print("Created results directory")

def run_jobs_in_batches(jobs, n_cores):
    """Run jobs in batches using multiprocessing Pool"""
    
    print(f"Running {len(jobs)} jobs in batches of {n_cores}")
    
    successful = 0
    failed = 0
    
    # Process jobs in batches
    for i in range(0, len(jobs), n_cores):
        batch = jobs[i:i + n_cores]
        print(f"\nBatch {i//n_cores + 1}: Running {len(batch)} jobs...")
        
        with Pool(processes=len(batch)) as pool:
            # Start all jobs in this batch
            results = []
            for cmd in batch:
                result = pool.apply_async(run_single_subprocess, (cmd,))
                results.append(result)
            
            # Wait for all jobs in this batch to complete
            batch_results = []
            for j, result in enumerate(results):
                try:
                    success, output = result.get(timeout=620)  # Slightly longer than subprocess timeout
                    if success:
                        successful += 1
                        print(f"  Job {i+j+1}: {output}")
                    else:
                        failed += 1
                        print(f"  Job {i+j+1}: FAILED - {output}")
                    batch_results.append((success, output))
                except Exception as e:
                    failed += 1
                    print(f"  Job {i+j+1}: ERROR - {e}")
                    batch_results.append((False, str(e)))
        
        print(f"Batch completed: {sum(1 for s, _ in batch_results if s)}/{len(batch)} successful")
    
    return successful, failed

def run_parallel_power_analysis(icc_results, baseline_prop=BASELINE_PROP, 
                               max_mice=MAX_MICE_TESTED, n_sims=N_SIMULATIONS,
                               relative_effects=RELATIVE_EFFECTS, 
                               cluster_sizes=CLUSTER_SIZES, 
                               target_power=TARGET_POWER, n_cores=N_CORES):
    """
    Run power analysis using subprocess calls in batches
    """
    
    # Get median ICC
    median_icc = get_median_icc(icc_results)
    
    print(f"Power Analysis Configuration:")
    print(f"  Median ICC: {median_icc:.4f}")
    print(f"  Baseline proportion: {baseline_prop}")
    print(f"  Target power: {target_power}")
    print(f"  Simulations per job: {n_sims:,}")
    print(f"  Max mice tested: {max_mice}")
    print(f"  Parallel processes: {n_cores}")
    print(f"  Effect sizes: {len(relative_effects)} levels")
    print(f"  Cluster sizes: {cluster_sizes}")
    
    # Clean results directory
    clean_results_directory()
    
    # Create job list
    jobs = create_job_list(baseline_prop, relative_effects, cluster_sizes, 
                          median_icc, target_power, max_mice, n_sims)
    
    print(f"  Total jobs: {len(jobs)}")
    print(f"  Estimated time: ~{len(jobs) * n_sims / (n_cores * 1000):.1f} minutes")
    
    # Run jobs
    start_time = time.time()
    successful, failed = run_jobs_in_batches(jobs, n_cores)
    elapsed_time = time.time() - start_time
    
    print(f"\nExecution completed in {elapsed_time/60:.1f} minutes")
    print(f"  Successful jobs: {successful}/{len(jobs)}")
    print(f"  Failed jobs: {failed}")
    
    # Collect results
    print("\nCollecting results...")
    all_results = []
    
    result_files = glob.glob('power_sim_results/result_*.csv')
    print(f"Found {len(result_files)} result files")
    
    for file in result_files:
        try:
            df = pd.read_csv(file)
            all_results.append(df)
        except Exception as e:
            print(f"Error reading {file}: {e}")
    
    if not all_results:
        print("No results found!")
        return None, None
    
    # Combine all results
    mice_required_df = pd.concat(all_results, ignore_index=True)
    mice_required_df = mice_required_df.sort_values(['cluster_size', 'relative_effect_pct'])
    
    print(f"Combined results: {len(mice_required_df)} rows")
    
    # Create plot
    plot_results(mice_required_df, baseline_prop, median_icc, target_power, n_sims)
    
    # Print summary
    print_summary(mice_required_df, target_power, max_mice, cluster_sizes)
    
    return mice_required_df, median_icc

def plot_results(mice_required_df, baseline_prop, median_icc, target_power, n_sims):
    """Create the power analysis plot"""
    fig, ax = plt.subplots(figsize=(9, 3))
    
    cluster_sizes = sorted(mice_required_df['cluster_size'].unique())
    colors = sns.color_palette("deep", len(cluster_sizes))
    
    for i, cluster_size in enumerate(cluster_sizes):
        data = mice_required_df[mice_required_df['cluster_size'] == cluster_size]
        ax.plot(data['relative_effect_pct'], data['required_mice_per_group'], 
                marker='o', linewidth=3, markersize=6, color=colors[i],
                label=f'{cluster_size} cells/mouse')
    
    ax.set_xlabel('Effect Size (% Reduction from Baseline)')
    ax.set_ylabel(f'Mice Required per Group (for {target_power*100}% Power)')
    ax.set_title(f'Sample Size Requirements vs Effect Size\n' + 
                f'({baseline_prop*100}% Baseline Disease Rate, ICC = {median_icc:.4f})')
    ax.grid(True, alpha=0.3)
    ax.legend(title='Sampling Strategy')
    ax.set_xlim(5, 80)
    ax.set_ylim(0, mice_required_df['required_mice_per_group'].max())
    
    # Add annotation
    annotation_text = (f'Target: {target_power*100}% Power\n'
                      f'Baseline: {baseline_prop*100}% diseased\n'
                      f'Simulations: {n_sims:,}\n'
                      f'Subprocess execution')
    
    ax.text(0.02, 0.98, annotation_text, transform=ax.transAxes, 
            fontsize=10, verticalalignment='top',
            bbox=dict(boxstyle='round', facecolor='lightgray', alpha=0.8))
    
    plt.tight_layout()
    plt.show()

def print_summary(mice_required_df, target_power, max_mice, cluster_sizes):
    """Print summary of results"""
    print(f"\nKey insights for {target_power*100}% power:")
    
    for cluster_size in cluster_sizes:
        if cluster_size in mice_required_df['cluster_size'].values:
            data = mice_required_df[mice_required_df['cluster_size'] == cluster_size]
            print(f"\n{cluster_size} cells/mouse:")
            
            for effect in [10, 20, 30, 50]:
                effect_data = data[data['relative_effect_pct'] == effect]
                if len(effect_data) > 0:
                    mice_needed = effect_data['required_mice_per_group'].iloc[0]
                    if mice_needed == max_mice:
                        print(f"  {effect}% effect: >{max_mice} mice per group")
                    else:
                        print(f"  {effect}% effect: {mice_needed} mice per group")

# =============================================================================
# USAGE:
# =============================================================================
mice_required_df, median_icc = run_parallel_power_analysis(icc_results)
```

```python
import pandas as pd
import numpy as np
import statsmodels.api as sm
from statsmodels.genmod import families
from statsmodels.genmod.cov_struct import Exchangeable
from tqdm import tqdm
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')

# =============================================================================
# CONFIGURABLE PARAMETERS
# =============================================================================

BASELINE_PROP = 0.25        # Baseline disease rate (25%)
MAX_MICE_TESTED = 100       # Maximum number of mice to test per group
N_SIMULATIONS = 10        # Number of simulations per power calculation
RELATIVE_EFFECTS = np.arange(0.05, 0.81, 0.05)  # 5% to 80% in 5% increments
CLUSTER_SIZES = [50, 100, 2500]                 # Cells per mouse
TARGET_POWER = 0.8          # Target statistical power (80%)

# =============================================================================

def simulate_single_experiment(baseline_prop, relative_effect, cluster_size, median_icc, n_clusters):
    """
    Simulate a single experiment and test for significance
    
    Returns:
    --------
    bool: True if p < 0.05, False otherwise
    """
    # Calculate treatment proportion
    treatment_prop = baseline_prop * (1 - relative_effect)
    treatment_prop = np.clip(treatment_prop, 0.001, 0.999)
    
    # Calculate logits
    baseline_logit = np.log(baseline_prop / (1 - baseline_prop))
    treatment_logit = np.log(treatment_prop / (1 - treatment_prop))
    
    # Generate random effects for each mouse
    control_effects = np.random.normal(0, np.sqrt(median_icc), n_clusters)
    treatment_effects = np.random.normal(0, np.sqrt(median_icc), n_clusters)
    
    # Calculate probabilities for each mouse
    control_logits = baseline_logit + control_effects
    treatment_logits = treatment_logit + treatment_effects
    
    control_probs = 1 / (1 + np.exp(-control_logits))
    treatment_probs = 1 / (1 + np.exp(-treatment_logits))
    
    control_probs = np.clip(control_probs, 0.001, 0.999)
    treatment_probs = np.clip(treatment_probs, 0.001, 0.999)
    
    # Generate data for each mouse
    y_values = []
    treatment_values = []
    cluster_ids = []
    
    # Control group
    for i in range(n_clusters):
        mouse_data = np.random.binomial(1, control_probs[i], cluster_size)
        y_values.extend(mouse_data)
        treatment_values.extend([0] * cluster_size)
        cluster_ids.extend([i] * cluster_size)
    
    # Treatment group  
    for i in range(n_clusters):
        mouse_data = np.random.binomial(1, treatment_probs[i], cluster_size)
        y_values.extend(mouse_data)
        treatment_values.extend([1] * cluster_size)
        cluster_ids.extend([i + n_clusters] * cluster_size)
    
    # Convert to arrays
    y = np.array(y_values)
    treatment = np.array(treatment_values)
    clusters = np.array(cluster_ids)
    
    # Check for valid data
    if len(np.unique(y)) <= 1:
        return False
    
    # Fit GEE model
    try:
        X = sm.add_constant(treatment)
        gee_model = sm.GEE(y, X, groups=clusters, 
                          cov_struct=Exchangeable(), 
                          family=families.Binomial())
        result = gee_model.fit(maxiter=50, disp=False)
        
        p_value = result.pvalues[1]
        return p_value < 0.05
        
    except:
        return False

def calculate_power(baseline_prop, relative_effect, cluster_size, median_icc, n_clusters, n_sims):
    """
    Calculate statistical power by running multiple simulations
    """
    significant_count = 0
    
    for _ in range(n_sims):
        if simulate_single_experiment(baseline_prop, relative_effect, cluster_size, median_icc, n_clusters):
            significant_count += 1
    
    return significant_count / n_sims

def find_required_mice(baseline_prop, relative_effect, cluster_size, median_icc, 
                      target_power=0.8, max_mice=100, n_sims=1000):
    """
    Find minimum number of mice needed using binary search
    """
    if relative_effect == 0:
        return max_mice
    
    low = 2
    high = max_mice
    best_mice = max_mice
    
    while low <= high:
        mid = (low + high) // 2
        
        power = calculate_power(baseline_prop, relative_effect, cluster_size, 
                               median_icc, mid, n_sims)
        
        if power >= target_power:
            best_mice = mid
            high = mid - 1
        else:
            low = mid + 1
    
    return best_mice

def run_power_analysis(icc_results, baseline_prop=BASELINE_PROP, max_mice=MAX_MICE_TESTED, 
                      n_sims=N_SIMULATIONS, relative_effects=RELATIVE_EFFECTS, 
                      cluster_sizes=CLUSTER_SIZES, target_power=TARGET_POWER):
    """
    Run complete power analysis
    """
    
    # Calculate median ICC
    all_icc_values = []
    for dataset in icc_results.keys():
        for treatment in icc_results[dataset].keys():
            if treatment not in ['overall', 'Wildtype']:
                for outcome in icc_results[dataset][treatment].keys():
                    icc_val = icc_results[dataset][treatment][outcome]
                    if icc_val is not None and not np.isnan(icc_val):
                        all_icc_values.append(icc_val)

    median_icc = np.median(all_icc_values)
    
    print(f"Using median ICC: {median_icc:.4f}")
    print(f"Running {n_sims:,} simulations per analysis")
    print(f"Maximum mice tested per group: {max_mice}")
    print(f"Target power: {target_power*100}%")
    print(f"Testing {len(relative_effects)} effect sizes Ã— {len(cluster_sizes)} cell counts")
    
    # Run analysis
    results = []
    total_combinations = len(relative_effects) * len(cluster_sizes)
    
    with tqdm(total=total_combinations, desc="Analyzing power") as pbar:
        for cluster_size in cluster_sizes:
            for relative_effect in relative_effects:
                
                required_mice = find_required_mice(
                    baseline_prop=baseline_prop,
                    relative_effect=relative_effect,
                    cluster_size=cluster_size,
                    median_icc=median_icc,
                    target_power=target_power,
                    max_mice=max_mice,
                    n_sims=n_sims
                )
                
                results.append({
                    'relative_effect_pct': relative_effect * 100,
                    'cluster_size': cluster_size,
                    'required_mice_per_group': required_mice
                })
                
                pbar.update(1)
    
    # Create DataFrame
    mice_required_df = pd.DataFrame(results)
    
    # Plot results
    fig, ax = plt.subplots(figsize=(9, 3))
    
    colors = sns.color_palette("deep", len(cluster_sizes))
    for i, cluster_size in enumerate(cluster_sizes):
        data = mice_required_df[mice_required_df['cluster_size'] == cluster_size]
        ax.plot(data['relative_effect_pct'], data['required_mice_per_group'], 
                marker='o', linewidth=3, markersize=6, color=colors[i],
                label=f'{cluster_size} cells/mouse')
    
    ax.set_xlabel('Effect Size (% Reduction from Baseline)')
    ax.set_ylabel(f'Mice Required per Group (for {target_power*100}% Power)')
    ax.set_title(f'Sample Size Requirements vs Effect Size\n' + 
                f'({baseline_prop*100}% Baseline Disease Rate, ICC = {median_icc:.4f})')
    ax.grid(True, alpha=0.3)
    ax.legend(title='Sampling Strategy')
    ax.set_xlim(5, 80)
    ax.set_ylim(0, max_mice)
    
    # Add annotation
    annotation_text = (f'Target: {target_power*100}% Power\n'
                      f'Baseline: {baseline_prop*100}% diseased\n'
                      f'Simulations: {n_sims:,}')
    
    ax.text(0.02, 0.98, annotation_text, transform=ax.transAxes, 
            fontsize=10, verticalalignment='top',
            bbox=dict(boxstyle='round', facecolor='lightgray', alpha=0.8))
    
    plt.tight_layout()
    plt.show()
    
    # Print summary
    print("\nPower analysis completed!")
    print(f"\nKey insights for {target_power*100}% power:")
    
    for cluster_size in cluster_sizes:
        data = mice_required_df[mice_required_df['cluster_size'] == cluster_size]
        print(f"\n{cluster_size} cells/mouse:")
        
        for effect in [10, 20, 30, 50]:
            effect_data = data[data['relative_effect_pct'] == effect]
            if len(effect_data) > 0:
                mice_needed = effect_data['required_mice_per_group'].iloc[0]
                if mice_needed == max_mice:
                    print(f"  {effect}% effect: >{max_mice} mice per group")
                else:
                    print(f"  {effect}% effect: {mice_needed} mice per group")
    
    return mice_required_df, median_icc

# =============================================================================
# USAGE:
# =============================================================================
mice_required_df, median_icc = run_power_analysis(icc_results)
```

```python
import pandas as pd
import numpy as np
from scipy import stats
import statsmodels.api as sm
from statsmodels.genmod import families
from statsmodels.genmod.cov_struct import Exchangeable
from tqdm import tqdm
import matplotlib.pyplot as plt
import seaborn as sns
from concurrent.futures import ThreadPoolExecutor, ProcessPoolExecutor
import multiprocessing as mp
import os

# =============================================================================
# CONFIGURABLE PARAMETERS - Modify these as needed
# =============================================================================

# Core simulation parameters
BASELINE_PROP = 0.25        # Baseline disease rate (25%)
MAX_MICE_TESTED = 100       # Maximum number of mice to test per group
N_SIMULATIONS = 10        # Number of simulations (recommend 10,000+ for publication)
USE_PARALLEL = True         # Enable parallel processing
N_CORES = 8                # Number of CPU cores to use (set to your preference)

# Effect size and cluster size ranges
RELATIVE_EFFECTS = np.arange(0.05, 0.81, 0.05)  # 5% to 80% in 5% increments
CLUSTER_SIZES = [50, 100, 2500]                 # Cells per mouse

# Analysis parameters
TARGET_POWER = 0.8          # Target statistical power (80%)

# =============================================================================

def fit_single_gee_model(sim_data):
    """
    Fit a single GEE model for one simulation
    Used for parallel processing
    
    Parameters:
    -----------
    sim_data : tuple
        (y, treatment, clusters) for one simulation
    
    Returns:
    --------
    bool : True if p < 0.05, False otherwise
    """
    y, treatment, clusters = sim_data
    
    try:
        X = sm.add_constant(treatment)
        gee_model = sm.GEE(y, X, groups=clusters, 
                         cov_struct=Exchangeable(), 
                         family=families.Binomial())
        gee_result = gee_model.fit()
        
        # Check if treatment effect is significant
        p_value = gee_result.pvalues[1]
        return p_value < 0.05
        
    except:
        return False

def simulate_clustered_binary_data_fixed_mice_parallel(baseline_prop, relative_effect, cluster_size, 
                                                     median_icc, n_clusters=5, n_sims=1000, 
                                                     use_parallel=True, n_cores=28):
    """
    Parallel vectorized calculation of power for fixed number of mice (clusters)
    
    Parameters:
    -----------
    baseline_prop : float
        Baseline disease proportion
    relative_effect : float
        Relative reduction in disease (0.2 = 20% reduction)
    cluster_size : int
        Number of cells per mouse
    median_icc : float
        Intracluster correlation coefficient
    n_clusters : int
        Number of mice per group
    n_sims : int
        Number of simulations to run
    use_parallel : bool
        Whether to use parallel processing
    n_cores : int
        Number of CPU cores to use
    """
    # Calculate treatment proportion from relative effect
    treatment_prop = baseline_prop * (1 - relative_effect)
    
    # Pre-calculate logits
    baseline_logit = np.log(baseline_prop / (1 - baseline_prop))
    treatment_logit = np.log(treatment_prop / (1 - treatment_prop))
    
    # Generate ALL random effects at once - Shape: (n_sims, n_clusters)
    cluster_effects_control = np.random.normal(0, np.sqrt(median_icc), (n_sims, n_clusters))
    cluster_effects_treatment = np.random.normal(0, np.sqrt(median_icc), (n_sims, n_clusters))
    
    # Calculate all cluster probabilities at once - Shape: (n_sims, n_clusters)
    control_logits = baseline_logit + cluster_effects_control
    treatment_logits = treatment_logit + cluster_effects_treatment
    
    control_probs = 1 / (1 + np.exp(-control_logits))
    treatment_probs = 1 / (1 + np.exp(-treatment_logits))
    
    # Clip probabilities
    control_probs = np.clip(control_probs, 0.001, 0.999)
    treatment_probs = np.clip(treatment_probs, 0.001, 0.999)
    
    # Generate ALL outcomes at once - Shape: (n_sims, n_clusters, cluster_size)
    control_outcomes = np.random.binomial(1, control_probs[:, :, np.newaxis], 
                                        (n_sims, n_clusters, cluster_size))
    treatment_outcomes = np.random.binomial(1, treatment_probs[:, :, np.newaxis], 
                                          (n_sims, n_clusters, cluster_size))
    
    # Prepare data for all simulations
    simulation_data = []
    for sim in range(n_sims):
        # Extract data for this simulation
        control_sim_data = control_outcomes[sim].flatten()
        treatment_sim_data = treatment_outcomes[sim].flatten()
        
        # Create cluster IDs
        control_cluster_ids = np.repeat([f'control_{i}' for i in range(n_clusters)], cluster_size)
        treatment_cluster_ids = np.repeat([f'treatment_{i}' for i in range(n_clusters)], cluster_size)
        
        # Combine data
        y = np.concatenate([control_sim_data, treatment_sim_data])
        treatment = np.concatenate([np.zeros(len(control_sim_data)), np.ones(len(treatment_sim_data))])
        clusters = np.concatenate([control_cluster_ids, treatment_cluster_ids])
        
        simulation_data.append((y, treatment, clusters))
    
    # Fit models in parallel or sequentially
    if use_parallel and n_cores > 1:
        # Use ProcessPoolExecutor for CPU-intensive tasks
        with ProcessPoolExecutor(max_workers=n_cores) as executor:
            # Process in chunks to show progress
            chunk_size = max(1, n_sims // 100)  # 100 progress updates max
            significant_results = []
            
            for i in range(0, n_sims, chunk_size):
                chunk = simulation_data[i:i + chunk_size]
                chunk_results = list(executor.map(fit_single_gee_model, chunk))
                significant_results.extend(chunk_results)
        
        significant_count = sum(significant_results)
        actual_sims_completed = len(significant_results)
        
    else:
        # Sequential processing with early stopping
        significant_count = 0
        actual_sims_completed = 0
        
        for sim, sim_data in enumerate(simulation_data):
            if fit_single_gee_model(sim_data):
                significant_count += 1
            
            actual_sims_completed += 1
            
            # Early stopping for extreme cases
            if sim >= 15:
                current_power = significant_count / actual_sims_completed
                if current_power == 0.0:
                    break
                elif current_power == 1.0 and sim >= 20:
                    break
    
    # Calculate final power
    power = significant_count / actual_sims_completed
    power = min(power, 1.0)
    return power

# Keep original function for backward compatibility
def simulate_clustered_binary_data_fixed_mice(baseline_prop, relative_effect, cluster_size, 
                                            median_icc, n_clusters=5, n_sims=1000):
    """
    Original non-vectorized version - kept for backward compatibility
    Uses parallel processing by default
    """
    return simulate_clustered_binary_data_fixed_mice_parallel(
        baseline_prop, relative_effect, cluster_size, median_icc, 
        n_clusters, n_sims, USE_PARALLEL, N_CORES
    )

def find_required_mice_for_effect_size(baseline_prop, relative_effect, cluster_size, 
                                     median_icc, target_power=0.8, max_mice=100, 
                                     n_sims=1000, use_parallel=True, n_cores=28):
    """
    Find minimum number of mice needed for given effect size using binary search
    
    Parameters:
    -----------
    baseline_prop : float
        Baseline disease proportion
    relative_effect : float
        Relative reduction in disease
    cluster_size : int
        Number of cells per mouse
    median_icc : float
        Intracluster correlation coefficient
    target_power : float
        Target statistical power (default 0.8)
    max_mice : int
        Maximum number of mice to test
    n_sims : int
        Number of simulations for power calculation
    use_parallel : bool
        Whether to use parallel processing
    n_cores : int
        Number of CPU cores to use
    """
    if relative_effect == 0:  # No effect, infinite mice needed
        return max_mice
    
    low = 2
    high = max_mice
    best_estimate = None
    
    # Binary search
    while low <= high:
        mid = (low + high) // 2
        
        power = simulate_clustered_binary_data_fixed_mice_parallel(
            baseline_prop=baseline_prop,
            relative_effect=relative_effect,
            cluster_size=cluster_size,
            median_icc=median_icc,
            n_clusters=mid,
            n_sims=n_sims,
            use_parallel=use_parallel,
            n_cores=n_cores
        )
        
        if abs(power - target_power) <= 0.1:  # Within 10% of target
            return mid
        elif power < target_power:
            low = mid + 1
        else:
            best_estimate = mid
            high = mid - 1
    
    return best_estimate if best_estimate is not None else max_mice

def run_power_analysis(icc_results, baseline_prop=BASELINE_PROP, max_mice=MAX_MICE_TESTED, 
                      n_sims=N_SIMULATIONS, relative_effects=RELATIVE_EFFECTS, 
                      cluster_sizes=CLUSTER_SIZES, target_power=TARGET_POWER,
                      use_parallel=USE_PARALLEL, n_cores=N_CORES):
    """
    Run complete power analysis with configurable parameters
    """
    
    # Get median ICC
    all_icc_values = []
    for dataset in icc_results.keys():
        for treatment in icc_results[dataset].keys():
            if treatment not in ['overall', 'Wildtype']:
                for outcome in icc_results[dataset][treatment].keys():
                    icc_val = icc_results[dataset][treatment][outcome]
                    if icc_val is not None and not np.isnan(icc_val):
                        all_icc_values.append(icc_val)

    median_icc = np.median(all_icc_values)
    print(f"Using median ICC: {median_icc:.4f}")
    print(f"Running {n_sims:,} simulations per analysis")
    print(f"Maximum mice tested per group: {max_mice}")
    if use_parallel:
        print(f"Using parallel processing with {n_cores} cores")
    else:
        print("Using sequential processing")
    
    print(f"Finding required mice for {target_power*100}% power across effect sizes")
    print(f"Testing {len(relative_effects)} effect sizes Ã— {len(cluster_sizes)} cell counts")

    # Run analysis
    results = []
    total_combinations = len(relative_effects) * len(cluster_sizes)

    with tqdm(total=total_combinations, desc="Finding required mice per effect size") as pbar:
        for cluster_size in cluster_sizes:
            for relative_effect in relative_effects:
                required_mice = find_required_mice_for_effect_size(
                    baseline_prop=baseline_prop,
                    relative_effect=relative_effect,
                    cluster_size=cluster_size,
                    median_icc=median_icc,
                    target_power=target_power,
                    max_mice=max_mice,
                    n_sims=n_sims,
                    use_parallel=use_parallel,
                    n_cores=n_cores
                )
                
                results.append({
                    'relative_effect_pct': relative_effect * 100,
                    'cluster_size': cluster_size,
                    'required_mice_per_group': required_mice
                })
                
                pbar.update(1)

    # Convert to DataFrame
    mice_required_df = pd.DataFrame(results)
    
    # Create the plot
    plt.figure(figsize=(9, 3))

    # Plot lines for each cell count
    colors = sns.color_palette("deep", len(cluster_sizes))
    for i, cluster_size in enumerate(cluster_sizes):
        cluster_data = mice_required_df[mice_required_df['cluster_size'] == cluster_size]
        
        plt.plot(cluster_data['relative_effect_pct'], cluster_data['required_mice_per_group'], 
                marker='o', linewidth=3, markersize=6, color=colors[i],
                label=f'{cluster_size} cells/mouse')

    plt.xlabel('Effect Size (% Reduction from Baseline)')
    ylabel_text = f'Mice Required per Group (for {target_power*100}% Power)'
    plt.ylabel(ylabel_text)
    title_text = f'Sample Size Requirements vs Effect Size\n({baseline_prop*100}% Baseline Disease Rate, ICC = {median_icc:.4f})'
    plt.title(title_text)
    plt.grid(True, alpha=0.3)
    plt.legend(title='Sampling Strategy')

    # Set axis limits
    plt.xlim(5, 80)
    plt.ylim(0, max_mice)

    # Add annotations
    parallel_text = f"Parallel ({n_cores} cores)" if use_parallel else "Sequential"
    target_text = f'Target: {target_power*100}% Power'
    baseline_text = f'Baseline: {baseline_prop*100}% diseased'
    sims_text = f'Sims: {n_sims:,}'
    mode_text = f'Mode: {parallel_text}'
    
    annotation_text = target_text + '\n' + baseline_text + '\n' + sims_text + '\n' + mode_text
    
    plt.text(0.02, 0.98, annotation_text, 
            transform=plt.gca().transAxes, fontsize=10, verticalalignment='top',
            bbox=dict(boxstyle='round', facecolor='lightgray', alpha=0.8))

    plt.tight_layout()
    plt.show()

    print("Sample size analysis completed!")
    print("Results stored in 'mice_required_df' DataFrame")

    # Show key insights
    power_text = f"\nKey insights for {target_power*100}% power:"
    print(power_text)
    for cluster_size in cluster_sizes:
        cluster_data = mice_required_df[mice_required_df['cluster_size'] == cluster_size]
        print(f"\n{cluster_size} cells/mouse:")
        for effect in [10, 20, 30, 50]:
            effect_data = cluster_data[cluster_data['relative_effect_pct'] == effect]
            if len(effect_data) > 0:
                mice_needed = effect_data['required_mice_per_group'].iloc[0]
                if mice_needed == max_mice:
                    print(f"  {effect}% effect: >{max_mice} mice per group (underpowered)")
                else:
                    print(f"  {effect}% effect: {mice_needed} mice per group")

    # Calculate efficiency gains
    if len(cluster_sizes) >= 2:
        max_cluster = max(cluster_sizes)
        min_cluster = min(cluster_sizes)
        efficiency_text = f"\nEfficiency comparison (mice saved by using {max_cluster} vs {min_cluster} cells):"
        print(efficiency_text)
        for effect in [10, 20, 30, 50]:
            mice_min_data = mice_required_df[(mice_required_df['cluster_size'] == min_cluster) & 
                                           (mice_required_df['relative_effect_pct'] == effect)]
            mice_max_data = mice_required_df[(mice_required_df['cluster_size'] == max_cluster) & 
                                           (mice_required_df['relative_effect_pct'] == effect)]
            
            if len(mice_min_data) > 0 and len(mice_max_data) > 0:
                mice_min = mice_min_data['required_mice_per_group'].iloc[0]
                mice_max = mice_max_data['required_mice_per_group'].iloc[0]
                
                if mice_min == max_mice:
                    effect_text = f"  {effect}% effect: {min_cluster} cells/mouse requires >{max_mice} mice (vs {mice_max} for {max_cluster} cells)"
                    print(effect_text)
                else:
                    savings = mice_min - mice_max
                    savings_text = f"  {effect}% effect: {savings} fewer mice needed ({mice_min} â†’ {mice_max})"
                    print(savings_text)
    
    return mice_required_df, median_icc

# =============================================================================
# USAGE EXAMPLE (uncomment to run):
# =============================================================================
# Assuming you have icc_results from previous analysis:
mice_required_df, median_icc = run_power_analysis(icc_results)
```

```python
import pandas as pd
import numpy as np
from scipy import stats
import statsmodels.api as sm
from statsmodels.genmod import families
from statsmodels.genmod.cov_struct import Exchangeable
from tqdm import tqdm
import matplotlib.pyplot as plt
import seaborn as sns
from concurrent.futures import ProcessPoolExecutor
import multiprocessing as mp
import os

# =============================================================================
# CONFIGURABLE PARAMETERS - Modify these as needed
# =============================================================================

# Core simulation parameters
BASELINE_PROP = 0.25        # Baseline disease rate (25%)
MAX_MICE_TESTED = 100       # Maximum number of mice to test per group
N_SIMULATIONS = 500         # Increased for better parallel efficiency
USE_PARALLEL = True         # Enable parallel processing
N_CORES = mp.cpu_count()    # Use all available cores

# Effect size and cluster size ranges
RELATIVE_EFFECTS = np.arange(0.05, 0.81, 0.05)  # 5% to 80% in 5% increments
CLUSTER_SIZES = [50, 100, 2500]                 # Cells per mouse

# Analysis parameters
TARGET_POWER = 0.8          # Target statistical power (80%)

# =============================================================================

def fit_gee_batch(params_batch):
    """
    Fit multiple GEE models in a single worker process to reduce overhead
    
    Parameters:
    -----------
    params_batch : list of tuples
        List of (baseline_prop, relative_effect, cluster_size, median_icc, n_clusters, sim_seeds)
    
    Returns:
    --------
    list of bool : Results for each simulation in the batch
    """
    # Import everything within the worker
    import numpy as np
    import statsmodels.api as sm
    from statsmodels.genmod import families
    from statsmodels.genmod.cov_struct import Exchangeable
    import warnings
    warnings.filterwarnings('ignore')
    
    results = []
    
    for baseline_prop, relative_effect, cluster_size, median_icc, n_clusters, sim_seed in params_batch:
        try:
            # Set random seed for this simulation
            np.random.seed(sim_seed)
            
            # Calculate treatment proportion from relative effect
            treatment_prop = baseline_prop * (1 - relative_effect)
            
            # Handle edge cases
            if treatment_prop <= 0.001:
                treatment_prop = 0.001
            if treatment_prop >= 0.999:
                treatment_prop = 0.999
                
            # Pre-calculate logits
            baseline_logit = np.log(baseline_prop / (1 - baseline_prop))
            treatment_logit = np.log(treatment_prop / (1 - treatment_prop))
            
            # Generate random effects for this simulation
            cluster_effects_control = np.random.normal(0, np.sqrt(median_icc), n_clusters)
            cluster_effects_treatment = np.random.normal(0, np.sqrt(median_icc), n_clusters)
            
            # Calculate cluster probabilities
            control_logits = baseline_logit + cluster_effects_control
            treatment_logits = treatment_logit + cluster_effects_treatment
            
            control_probs = 1 / (1 + np.exp(-control_logits))
            treatment_probs = 1 / (1 + np.exp(-treatment_logits))
            
            # Clip probabilities
            control_probs = np.clip(control_probs, 0.001, 0.999)
            treatment_probs = np.clip(treatment_probs, 0.001, 0.999)
            
            # Generate outcomes for each cluster - vectorized
            control_outcomes = []
            treatment_outcomes = []
            
            for i in range(n_clusters):
                control_cluster_data = np.random.binomial(1, control_probs[i], cluster_size)
                treatment_cluster_data = np.random.binomial(1, treatment_probs[i], cluster_size)
                control_outcomes.extend(control_cluster_data)
                treatment_outcomes.extend(treatment_cluster_data)
            
            # Create cluster IDs - use integers for efficiency
            control_cluster_ids = np.repeat(np.arange(n_clusters), cluster_size)
            treatment_cluster_ids = np.repeat(np.arange(n_clusters, 2*n_clusters), cluster_size)
            
            # Combine data
            y = np.array(control_outcomes + treatment_outcomes, dtype=np.float64)
            treatment = np.concatenate([np.zeros(len(control_outcomes)), np.ones(len(treatment_outcomes))])
            clusters = np.concatenate([control_cluster_ids, treatment_cluster_ids])
            
            # Basic data validation
            if len(y) == 0 or len(np.unique(y)) == 1 or np.var(y) == 0:
                results.append(False)
                continue
                
            # Fit GEE model with minimal overhead
            X = np.column_stack([np.ones(len(y)), treatment])
            
            gee_model = sm.GEE(y, X, groups=clusters, 
                             cov_struct=Exchangeable(), 
                             family=families.Binomial())
            
            # Fit with reduced iterations for speed
            gee_result = gee_model.fit(maxiter=25, tol=1e-4, disp=False)
            
            # Check significance
            p_value = gee_result.pvalues[1]
            results.append(p_value < 0.05)
            
        except Exception:
            results.append(False)
    
    return results

def simulate_clustered_binary_data_fixed_mice_parallel(baseline_prop, relative_effect, cluster_size, 
                                                     median_icc, n_clusters=5, n_sims=1000, 
                                                     use_parallel=True, n_cores=None):
    """
    Parallel calculation of power for fixed number of mice (clusters)
    """
    if n_cores is None:
        n_cores = N_CORES
        
    # Prepare simulation parameters
    base_seed = np.random.randint(0, 1000000)
    sim_seeds = [base_seed + i for i in range(n_sims)]
    
    if use_parallel and n_cores > 1:
        # Create batches for parallel processing - larger batches reduce overhead
        batch_size = max(1, n_sims // (n_cores * 4))  # 4 batches per core
        batches = []
        
        for i in range(0, n_sims, batch_size):
            batch_seeds = sim_seeds[i:i + batch_size]
            batch = [(baseline_prop, relative_effect, cluster_size, median_icc, n_clusters, seed) 
                    for seed in batch_seeds]
            batches.append(batch)
        
        # Process batches in parallel
        with ProcessPoolExecutor(max_workers=n_cores) as executor:
            batch_results = list(executor.map(fit_gee_batch, batches))
        
        # Flatten results
        significant_results = []
        for batch_result in batch_results:
            significant_results.extend(batch_result)
            
        significant_count = sum(significant_results)
        actual_sims_completed = len(significant_results)
        
    else:
        # Sequential processing
        all_params = [(baseline_prop, relative_effect, cluster_size, median_icc, n_clusters, seed) 
                     for seed in sim_seeds]
        significant_results = fit_gee_batch(all_params)
        significant_count = sum(significant_results)
        actual_sims_completed = len(significant_results)
    
    # Calculate final power
    power = significant_count / actual_sims_completed if actual_sims_completed > 0 else 0.0
    return min(power, 1.0)

def find_required_mice_for_effect_size(baseline_prop, relative_effect, cluster_size, 
                                     median_icc, target_power=0.8, max_mice=100, 
                                     n_sims=1000, use_parallel=True, n_cores=None):
    """
    Find minimum number of mice needed for given effect size using binary search
    """
    if n_cores is None:
        n_cores = N_CORES
        
    if relative_effect == 0:  # No effect case
        return max_mice
    
    # Quick check if even max_mice won't work
    if relative_effect < 0.01:  # Very small effect
        return max_mice
    
    low = 2
    high = max_mice
    best_estimate = None
    tolerance = 0.05  # 5% tolerance around target power
    
    # Binary search with early termination
    iterations = 0
    max_iterations = 10  # Limit binary search iterations
    
    while low <= high and iterations < max_iterations:
        mid = (low + high) // 2
        iterations += 1
        
        try:
            power = simulate_clustered_binary_data_fixed_mice_parallel(
                baseline_prop=baseline_prop,
                relative_effect=relative_effect,
                cluster_size=cluster_size,
                median_icc=median_icc,
                n_clusters=mid,
                n_sims=n_sims,
                use_parallel=use_parallel,
                n_cores=n_cores
            )
            
            if abs(power - target_power) <= tolerance:
                return mid
            elif power < target_power:
                low = mid + 1
            else:
                best_estimate = mid
                high = mid - 1
                
        except Exception:
            high = mid - 1
    
    return best_estimate if best_estimate is not None else max_mice

def run_power_analysis(icc_results, baseline_prop=BASELINE_PROP, max_mice=MAX_MICE_TESTED, 
                      n_sims=N_SIMULATIONS, relative_effects=RELATIVE_EFFECTS, 
                      cluster_sizes=CLUSTER_SIZES, target_power=TARGET_POWER,
                      use_parallel=USE_PARALLEL, n_cores=N_CORES):
    """
    Run complete power analysis with optimized parallel processing
    """
    
    # Get median ICC
    all_icc_values = []
    for dataset in icc_results.keys():
        for treatment in icc_results[dataset].keys():
            if treatment not in ['overall', 'Wildtype']:
                for outcome in icc_results[dataset][treatment].keys():
                    icc_val = icc_results[dataset][treatment][outcome]
                    if icc_val is not None and not np.isnan(icc_val):
                        all_icc_values.append(icc_val)

    median_icc = np.median(all_icc_values)
    print(f"Using median ICC: {median_icc:.4f}")
    print(f"Running {n_sims:,} simulations per analysis")
    print(f"Maximum mice tested per group: {max_mice}")
    if use_parallel:
        print(f"Using ProcessPoolExecutor with {n_cores} cores")
        print(f"Batch processing enabled for efficiency")
    else:
        print("Using sequential processing")
    
    print(f"Finding required mice for {target_power*100}% power across effect sizes")
    print(f"Testing {len(relative_effects)} effect sizes Ã— {len(cluster_sizes)} cell counts")

    # Run analysis
    results = []
    total_combinations = len(relative_effects) * len(cluster_sizes)

    with tqdm(total=total_combinations, desc="Power analysis", 
              bar_format='{l_bar}{bar}| {n_fmt}/{total_fmt} [{elapsed}<{remaining}]') as pbar:
        
        for cluster_size in cluster_sizes:
            for relative_effect in relative_effects:
                try:
                    required_mice = find_required_mice_for_effect_size(
                        baseline_prop=baseline_prop,
                        relative_effect=relative_effect,
                        cluster_size=cluster_size,
                        median_icc=median_icc,
                        target_power=target_power,
                        max_mice=max_mice,
                        n_sims=n_sims,
                        use_parallel=use_parallel,
                        n_cores=n_cores
                    )
                except Exception as e:
                    required_mice = max_mice
                
                results.append({
                    'relative_effect_pct': relative_effect * 100,
                    'cluster_size': cluster_size,
                    'required_mice_per_group': required_mice
                })
                
                pbar.update(1)

    # Convert to DataFrame and create plot
    mice_required_df = pd.DataFrame(results)
    
    # Create the plot
    fig, ax = plt.subplots(figsize=(9, 3))

    # Plot lines for each cell count
    colors = sns.color_palette("deep", len(cluster_sizes))
    for i, cluster_size in enumerate(cluster_sizes):
        cluster_data = mice_required_df[mice_required_df['cluster_size'] == cluster_size]
        
        ax.plot(cluster_data['relative_effect_pct'], cluster_data['required_mice_per_group'], 
                marker='o', linewidth=3, markersize=6, color=colors[i],
                label=f'{cluster_size} cells/mouse')

    ax.set_xlabel('Effect Size (% Reduction from Baseline)')
    ax.set_ylabel(f'Mice Required per Group (for {target_power*100}% Power)')
    ax.set_title(f'Sample Size Requirements vs Effect Size\n({baseline_prop*100}% Baseline Disease Rate, ICC = {median_icc:.4f})')
    ax.grid(True, alpha=0.3)
    ax.legend(title='Sampling Strategy')
    ax.set_xlim(5, 80)
    ax.set_ylim(0, max_mice)

    # Add annotations
    parallel_text = f"ProcessPool ({n_cores} cores)" if use_parallel else "Sequential"
    annotation_text = f'Target: {target_power*100}% Power\nBaseline: {baseline_prop*100}% diseased\nSims: {n_sims:,}\nMode: {parallel_text}'
    
    ax.text(0.02, 0.98, annotation_text, 
            transform=ax.transAxes, fontsize=10, verticalalignment='top',
            bbox=dict(boxstyle='round', facecolor='lightgray', alpha=0.8))

    plt.tight_layout()
    plt.show()

    print("\nSample size analysis completed!")
    print("Results stored in 'mice_required_df' DataFrame")

    # Show key insights
    print(f"\nKey insights for {target_power*100}% power:")
    for cluster_size in cluster_sizes:
        cluster_data = mice_required_df[mice_required_df['cluster_size'] == cluster_size]
        print(f"\n{cluster_size} cells/mouse:")
        for effect in [10, 20, 30, 50]:
            effect_data = cluster_data[cluster_data['relative_effect_pct'] == effect]
            if len(effect_data) > 0:
                mice_needed = effect_data['required_mice_per_group'].iloc[0]
                if mice_needed == max_mice:
                    print(f"  {effect}% effect: >{max_mice} mice per group (underpowered)")
                else:
                    print(f"  {effect}% effect: {mice_needed} mice per group")

    return mice_required_df, median_icc

# =============================================================================
# USAGE EXAMPLE:
# =============================================================================
mice_required_df, median_icc = run_power_analysis(icc_results)
```

```python
import pandas as pd
import numpy as np
from scipy import stats
import statsmodels.api as sm
from statsmodels.genmod import families
from statsmodels.genmod.cov_struct import Exchangeable
from tqdm import tqdm
import matplotlib.pyplot as plt
import seaborn as sns
from concurrent.futures import ThreadPoolExecutor
import multiprocessing as mp
import os

# =============================================================================
# CONFIGURABLE PARAMETERS - Modify these as needed
# =============================================================================

# Core simulation parameters
BASELINE_PROP = 0.25        # Baseline disease rate (25%)
MAX_MICE_TESTED = 100       # Maximum number of mice to test per group
N_SIMULATIONS = 100         # Increased from 10 for better stability
USE_PARALLEL = True         # Enable parallel processing
N_CORES = mp.cpu_count()    # Use all available cores

# Effect size and cluster size ranges
RELATIVE_EFFECTS = np.arange(0.05, 0.81, 0.05)  # 5% to 80% in 5% increments
CLUSTER_SIZES = [50, 100, 2500]                 # Cells per mouse

# Analysis parameters
TARGET_POWER = 0.8          # Target statistical power (80%)

# =============================================================================

def fit_single_gee_model(sim_data):
    """
    Fit a single GEE model for one simulation
    Used for parallel processing
    
    Parameters:
    -----------
    sim_data : tuple
        (y, treatment, clusters) for one simulation
    
    Returns:
    --------
    bool : True if p < 0.05, False otherwise
    """
    y, treatment, clusters = sim_data
    
    try:
        # Add some basic data validation
        if len(y) == 0 or len(set(y)) == 1:  # No data or no variation
            return False
            
        X = sm.add_constant(treatment)
        gee_model = sm.GEE(y, X, groups=clusters, 
                         cov_struct=Exchangeable(), 
                         family=families.Binomial())
        
        # Add convergence parameters
        gee_result = gee_model.fit(maxiter=50, tol=1e-6, disp=False)
        
        # Check if model converged and treatment effect is significant
        if hasattr(gee_result, 'converged') and not gee_result.converged:
            return False
            
        p_value = gee_result.pvalues[1]
        return p_value < 0.05
        
    except Exception as e:
        # Silently return False for any fitting issues
        return False

def simulate_clustered_binary_data_fixed_mice_parallel(baseline_prop, relative_effect, cluster_size, 
                                                     median_icc, n_clusters=5, n_sims=1000, 
                                                     use_parallel=True, n_cores=None):
    """
    Parallel vectorized calculation of power for fixed number of mice (clusters)
    
    Parameters:
    -----------
    baseline_prop : float
        Baseline disease proportion
    relative_effect : float
        Relative reduction in disease (0.2 = 20% reduction)
    cluster_size : int
        Number of cells per mouse
    median_icc : float
        Intracluster correlation coefficient
    n_clusters : int
        Number of mice per group
    n_sims : int
        Number of simulations to run
    use_parallel : bool
        Whether to use parallel processing
    n_cores : int
        Number of CPU cores to use
    """
    if n_cores is None:
        n_cores = N_CORES
        
    # Calculate treatment proportion from relative effect
    treatment_prop = baseline_prop * (1 - relative_effect)
    
    # Pre-calculate logits
    baseline_logit = np.log(baseline_prop / (1 - baseline_prop))
    treatment_logit = np.log(treatment_prop / (1 - treatment_prop))
    
    # Generate ALL random effects at once - Shape: (n_sims, n_clusters)
    np.random.seed(42)  # For reproducibility
    cluster_effects_control = np.random.normal(0, np.sqrt(median_icc), (n_sims, n_clusters))
    cluster_effects_treatment = np.random.normal(0, np.sqrt(median_icc), (n_sims, n_clusters))
    
    # Calculate all cluster probabilities at once - Shape: (n_sims, n_clusters)
    control_logits = baseline_logit + cluster_effects_control
    treatment_logits = treatment_logit + cluster_effects_treatment
    
    control_probs = 1 / (1 + np.exp(-control_logits))
    treatment_probs = 1 / (1 + np.exp(-treatment_logits))
    
    # Clip probabilities
    control_probs = np.clip(control_probs, 0.001, 0.999)
    treatment_probs = np.clip(treatment_probs, 0.001, 0.999)
    
    # Generate ALL outcomes at once - Shape: (n_sims, n_clusters, cluster_size)
    control_outcomes = np.random.binomial(1, control_probs[:, :, np.newaxis], 
                                        (n_sims, n_clusters, cluster_size))
    treatment_outcomes = np.random.binomial(1, treatment_probs[:, :, np.newaxis], 
                                          (n_sims, n_clusters, cluster_size))
    
    # Prepare data for all simulations
    simulation_data = []
    for sim in range(n_sims):
        # Extract data for this simulation
        control_sim_data = control_outcomes[sim].flatten()
        treatment_sim_data = treatment_outcomes[sim].flatten()
        
        # Create cluster IDs
        control_cluster_ids = np.repeat([f'c_{i}' for i in range(n_clusters)], cluster_size)
        treatment_cluster_ids = np.repeat([f't_{i}' for i in range(n_clusters)], cluster_size)
        
        # Combine data
        y = np.concatenate([control_sim_data, treatment_sim_data])
        treatment = np.concatenate([np.zeros(len(control_sim_data)), np.ones(len(treatment_sim_data))])
        clusters = np.concatenate([control_cluster_ids, treatment_cluster_ids])
        
        simulation_data.append((y, treatment, clusters))
    
    # Fit models in parallel or sequentially
    if use_parallel and n_cores > 1:
        # Use ThreadPoolExecutor instead of ProcessPoolExecutor
        with ThreadPoolExecutor(max_workers=n_cores) as executor:
            significant_results = list(executor.map(fit_single_gee_model, simulation_data))
        
        significant_count = sum(significant_results)
        actual_sims_completed = len(significant_results)
        
    else:
        # Sequential processing
        significant_count = 0
        actual_sims_completed = 0
        
        for sim, sim_data in enumerate(simulation_data):
            if fit_single_gee_model(sim_data):
                significant_count += 1
            
            actual_sims_completed += 1
            
            # Early stopping for extreme cases (only in sequential mode)
            if sim >= 20:
                current_power = significant_count / actual_sims_completed
                if current_power == 0.0 and sim >= 50:
                    break
                elif current_power == 1.0 and sim >= 30:
                    break
    
    # Calculate final power
    power = significant_count / actual_sims_completed if actual_sims_completed > 0 else 0.0
    power = min(power, 1.0)
    return power

# Keep original function for backward compatibility
def simulate_clustered_binary_data_fixed_mice(baseline_prop, relative_effect, cluster_size, 
                                            median_icc, n_clusters=5, n_sims=1000):
    """
    Original non-vectorized version - kept for backward compatibility
    Uses parallel processing by default
    """
    return simulate_clustered_binary_data_fixed_mice_parallel(
        baseline_prop, relative_effect, cluster_size, median_icc, 
        n_clusters, n_sims, USE_PARALLEL, N_CORES
    )

def find_required_mice_for_effect_size(baseline_prop, relative_effect, cluster_size, 
                                     median_icc, target_power=0.8, max_mice=100, 
                                     n_sims=1000, use_parallel=True, n_cores=None):
    """
    Find minimum number of mice needed for given effect size using binary search
    
    Parameters:
    -----------
    baseline_prop : float
        Baseline disease proportion
    relative_effect : float
        Relative reduction in disease
    cluster_size : int
        Number of cells per mouse
    median_icc : float
        Intracluster correlation coefficient
    target_power : float
        Target statistical power (default 0.8)
    max_mice : int
        Maximum number of mice to test
    n_sims : int
        Number of simulations for power calculation
    use_parallel : bool
        Whether to use parallel processing
    n_cores : int
        Number of CPU cores to use
    """
    if n_cores is None:
        n_cores = N_CORES
        
    if relative_effect == 0:  # No effect, infinite mice needed
        return max_mice
    
    low = 2
    high = max_mice
    best_estimate = None
    
    # Binary search
    while low <= high:
        mid = (low + high) // 2
        
        try:
            power = simulate_clustered_binary_data_fixed_mice_parallel(
                baseline_prop=baseline_prop,
                relative_effect=relative_effect,
                cluster_size=cluster_size,
                median_icc=median_icc,
                n_clusters=mid,
                n_sims=n_sims,
                use_parallel=use_parallel,
                n_cores=n_cores
            )
            
            if abs(power - target_power) <= 0.05:  # Within 5% of target (tighter tolerance)
                return mid
            elif power < target_power:
                low = mid + 1
            else:
                best_estimate = mid
                high = mid - 1
                
        except Exception as e:
            # If simulation fails, try with fewer mice
            high = mid - 1
    
    return best_estimate if best_estimate is not None else max_mice

def run_power_analysis(icc_results, baseline_prop=BASELINE_PROP, max_mice=MAX_MICE_TESTED, 
                      n_sims=N_SIMULATIONS, relative_effects=RELATIVE_EFFECTS, 
                      cluster_sizes=CLUSTER_SIZES, target_power=TARGET_POWER,
                      use_parallel=USE_PARALLEL, n_cores=N_CORES):
    """
    Run complete power analysis with configurable parameters
    """
    
    # Get median ICC
    all_icc_values = []
    for dataset in icc_results.keys():
        for treatment in icc_results[dataset].keys():
            if treatment not in ['overall', 'Wildtype']:
                for outcome in icc_results[dataset][treatment].keys():
                    icc_val = icc_results[dataset][treatment][outcome]
                    if icc_val is not None and not np.isnan(icc_val):
                        all_icc_values.append(icc_val)

    median_icc = np.median(all_icc_values)
    print(f"Using median ICC: {median_icc:.4f}")
    print(f"Running {n_sims:,} simulations per analysis")
    print(f"Maximum mice tested per group: {max_mice}")
    if use_parallel:
        print(f"Using ThreadPoolExecutor with {n_cores} cores")
    else:
        print("Using sequential processing")
    
    print(f"Finding required mice for {target_power*100}% power across effect sizes")
    print(f"Testing {len(relative_effects)} effect sizes Ã— {len(cluster_sizes)} cell counts")

    # Run analysis
    results = []
    total_combinations = len(relative_effects) * len(cluster_sizes)

    with tqdm(total=total_combinations, desc="Finding required mice per effect size") as pbar:
        for cluster_size in cluster_sizes:
            for relative_effect in relative_effects:
                try:
                    required_mice = find_required_mice_for_effect_size(
                        baseline_prop=baseline_prop,
                        relative_effect=relative_effect,
                        cluster_size=cluster_size,
                        median_icc=median_icc,
                        target_power=target_power,
                        max_mice=max_mice,
                        n_sims=n_sims,
                        use_parallel=use_parallel,
                        n_cores=n_cores
                    )
                except Exception as e:
                    print(f"Error for effect {relative_effect*100}%, cluster {cluster_size}: {e}")
                    required_mice = max_mice
                
                results.append({
                    'relative_effect_pct': relative_effect * 100,
                    'cluster_size': cluster_size,
                    'required_mice_per_group': required_mice
                })
                
                pbar.update(1)

    # Convert to DataFrame
    mice_required_df = pd.DataFrame(results)
    
    # Create the plot
    fig, ax = plt.subplots(figsize=(9, 3))

    # Plot lines for each cell count
    colors = sns.color_palette("deep", len(cluster_sizes))
    for i, cluster_size in enumerate(cluster_sizes):
        cluster_data = mice_required_df[mice_required_df['cluster_size'] == cluster_size]
        
        ax.plot(cluster_data['relative_effect_pct'], cluster_data['required_mice_per_group'], 
                marker='o', linewidth=3, markersize=6, color=colors[i],
                label=f'{cluster_size} cells/mouse')

    ax.set_xlabel('Effect Size (% Reduction from Baseline)')
    ylabel_text = f'Mice Required per Group (for {target_power*100}% Power)'
    ax.set_ylabel(ylabel_text)
    title_text = f'Sample Size Requirements vs Effect Size\n({baseline_prop*100}% Baseline Disease Rate, ICC = {median_icc:.4f})'
    ax.set_title(title_text)
    ax.grid(True, alpha=0.3)
    ax.legend(title='Sampling Strategy')

    # Set axis limits
    ax.set_xlim(5, 80)
    ax.set_ylim(0, max_mice)

    # Add annotations
    parallel_text = f"ThreadPool ({n_cores} cores)" if use_parallel else "Sequential"
    target_text = f'Target: {target_power*100}% Power'
    baseline_text = f'Baseline: {baseline_prop*100}% diseased'
    sims_text = f'Sims: {n_sims:,}'
    mode_text = f'Mode: {parallel_text}'
    
    annotation_text = target_text + '\n' + baseline_text + '\n' + sims_text + '\n' + mode_text
    
    ax.text(0.02, 0.98, annotation_text, 
            transform=ax.transAxes, fontsize=10, verticalalignment='top',
            bbox=dict(boxstyle='round', facecolor='lightgray', alpha=0.8))

    plt.tight_layout()
    plt.show()

    print("Sample size analysis completed!")
    print("Results stored in 'mice_required_df' DataFrame")

    # Show key insights
    power_text = f"\nKey insights for {target_power*100}% power:"
    print(power_text)
    for cluster_size in cluster_sizes:
        cluster_data = mice_required_df[mice_required_df['cluster_size'] == cluster_size]
        print(f"\n{cluster_size} cells/mouse:")
        for effect in [10, 20, 30, 50]:
            effect_data = cluster_data[cluster_data['relative_effect_pct'] == effect]
            if len(effect_data) > 0:
                mice_needed = effect_data['required_mice_per_group'].iloc[0]
                if mice_needed == max_mice:
                    print(f"  {effect}% effect: >{max_mice} mice per group (underpowered)")
                else:
                    print(f"  {effect}% effect: {mice_needed} mice per group")

    # Calculate efficiency gains
    if len(cluster_sizes) >= 2:
        max_cluster = max(cluster_sizes)
        min_cluster = min(cluster_sizes)
        efficiency_text = f"\nEfficiency comparison (mice saved by using {max_cluster} vs {min_cluster} cells):"
        print(efficiency_text)
        for effect in [10, 20, 30, 50]:
            mice_min_data = mice_required_df[(mice_required_df['cluster_size'] == min_cluster) & 
                                           (mice_required_df['relative_effect_pct'] == effect)]
            mice_max_data = mice_required_df[(mice_required_df['cluster_size'] == max_cluster) & 
                                           (mice_required_df['relative_effect_pct'] == effect)]
            
            if len(mice_min_data) > 0 and len(mice_max_data) > 0:
                mice_min = mice_min_data['required_mice_per_group'].iloc[0]
                mice_max = mice_max_data['required_mice_per_group'].iloc[0]
                
                if mice_min == max_mice:
                    effect_text = f"  {effect}% effect: {min_cluster} cells/mouse requires >{max_mice} mice (vs {mice_max} for {max_cluster} cells)"
                    print(effect_text)
                else:
                    savings = mice_min - mice_max
                    savings_text = f"  {effect}% effect: {savings} fewer mice needed ({mice_min} â†’ {mice_max})"
                    print(savings_text)
    
    return mice_required_df, median_icc

# =============================================================================
# USAGE EXAMPLE (uncomment to run):
# =============================================================================
# Assuming you have icc_results from previous analysis:
mice_required_df, median_icc = run_power_analysis(icc_results)
```

```python
import pandas as pd
import numpy as np
from scipy import stats
import statsmodels.api as sm
from statsmodels.genmod import families
from statsmodels.genmod.cov_struct import Exchangeable
from tqdm import tqdm
import matplotlib.pyplot as plt
import seaborn as sns
from concurrent.futures import ThreadPoolExecutor, ProcessPoolExecutor
import multiprocessing as mp
import os

# =============================================================================
# CONFIGURABLE PARAMETERS - Modify these as needed
# =============================================================================

# Core simulation parameters
BASELINE_PROP = 0.25        # Baseline disease rate (25%)
MAX_MICE_TESTED = 100       # Maximum number of mice to test per group
N_SIMULATIONS = 100         # Increased from 10 for better stability
USE_PARALLEL = True         # Enable parallel processing
N_CORES = mp.cpu_count()    # Use all available cores

# Effect size and cluster size ranges
RELATIVE_EFFECTS = np.arange(0.05, 0.81, 0.05)  # 5% to 80% in 5% increments
CLUSTER_SIZES = [50, 100, 2500]                 # Cells per mouse

# Analysis parameters
TARGET_POWER = 0.8          # Target statistical power (80%)

# =============================================================================

def fit_single_gee_model(sim_params):
    """
    Fit a single GEE model for one simulation
    Used for parallel processing
    
    Parameters:
    -----------
    sim_params : tuple
        (sim_index, baseline_prop, relative_effect, cluster_size, median_icc, n_clusters, random_seed)
    
    Returns:
    --------
    bool : True if p < 0.05, False otherwise
    """
    import numpy as np
    import pandas as pd
    import statsmodels.api as sm
    from statsmodels.genmod import families
    from statsmodels.genmod.cov_struct import Exchangeable
    
    sim_index, baseline_prop, relative_effect, cluster_size, median_icc, n_clusters, random_seed = sim_params
    
    try:
        # Set random seed for this simulation
        np.random.seed(random_seed + sim_index)
        
        # Calculate treatment proportion from relative effect
        treatment_prop = baseline_prop * (1 - relative_effect)
        
        # Pre-calculate logits
        baseline_logit = np.log(baseline_prop / (1 - baseline_prop))
        treatment_logit = np.log(treatment_prop / (1 - treatment_prop))
        
        # Generate random effects for this simulation
        cluster_effects_control = np.random.normal(0, np.sqrt(median_icc), n_clusters)
        cluster_effects_treatment = np.random.normal(0, np.sqrt(median_icc), n_clusters)
        
        # Calculate cluster probabilities
        control_logits = baseline_logit + cluster_effects_control
        treatment_logits = treatment_logit + cluster_effects_treatment
        
        control_probs = 1 / (1 + np.exp(-control_logits))
        treatment_probs = 1 / (1 + np.exp(-treatment_logits))
        
        # Clip probabilities
        control_probs = np.clip(control_probs, 0.001, 0.999)
        treatment_probs = np.clip(treatment_probs, 0.001, 0.999)
        
        # Generate outcomes for each cluster
        control_outcomes = []
        treatment_outcomes = []
        
        for i in range(n_clusters):
            control_cluster_data = np.random.binomial(1, control_probs[i], cluster_size)
            treatment_cluster_data = np.random.binomial(1, treatment_probs[i], cluster_size)
            control_outcomes.extend(control_cluster_data)
            treatment_outcomes.extend(treatment_cluster_data)
        
        # Create cluster IDs
        control_cluster_ids = np.repeat([f'c_{i}' for i in range(n_clusters)], cluster_size)
        treatment_cluster_ids = np.repeat([f't_{i}' for i in range(n_clusters)], cluster_size)
        
        # Combine data
        y = np.concatenate([control_outcomes, treatment_outcomes])
        treatment = np.concatenate([np.zeros(len(control_outcomes)), np.ones(len(treatment_outcomes))])
        clusters = np.concatenate([control_cluster_ids, treatment_cluster_ids])
        
        # Basic data validation
        if len(y) == 0 or len(set(y)) == 1:  # No data or no variation
            return False
            
        X = sm.add_constant(treatment)
        gee_model = sm.GEE(y, X, groups=clusters, 
                         cov_struct=Exchangeable(), 
                         family=families.Binomial())
        
        # Add convergence parameters
        gee_result = gee_model.fit(maxiter=50, tol=1e-6, disp=False)
        
        # Check if model converged and treatment effect is significant
        if hasattr(gee_result, 'converged') and not gee_result.converged:
            return False
            
        p_value = gee_result.pvalues[1]
        return p_value < 0.05
        
    except Exception as e:
        # Silently return False for any fitting issues
        return False

def simulate_clustered_binary_data_fixed_mice_parallel(baseline_prop, relative_effect, cluster_size, 
                                                     median_icc, n_clusters=5, n_sims=1000, 
                                                     use_parallel=True, n_cores=None):
    """
    Parallel calculation of power for fixed number of mice (clusters)
    
    Parameters:
    -----------
    baseline_prop : float
        Baseline disease proportion
    relative_effect : float
        Relative reduction in disease (0.2 = 20% reduction)
    cluster_size : int
        Number of cells per mouse
    median_icc : float
        Intracluster correlation coefficient
    n_clusters : int
        Number of mice per group
    n_sims : int
        Number of simulations to run
    use_parallel : bool
        Whether to use parallel processing
    n_cores : int
        Number of CPU cores to use
    """
    if n_cores is None:
        n_cores = N_CORES
        
    # Prepare simulation parameters for each simulation
    base_seed = 42
    simulation_params = [
        (sim_idx, baseline_prop, relative_effect, cluster_size, median_icc, n_clusters, base_seed)
        for sim_idx in range(n_sims)
    ]
    
    # Fit models in parallel or sequentially
    if use_parallel and n_cores > 1:
        # Use ProcessPoolExecutor for CPU-intensive tasks
        with ProcessPoolExecutor(max_workers=n_cores) as executor:
            significant_results = list(executor.map(fit_single_gee_model, simulation_params))
        
        significant_count = sum(significant_results)
        actual_sims_completed = len(significant_results)
        
    else:
        # Sequential processing
        significant_count = 0
        actual_sims_completed = 0
        
        for sim_params in simulation_params:
            if fit_single_gee_model(sim_params):
                significant_count += 1
            
            actual_sims_completed += 1
            
            # Early stopping for extreme cases (only in sequential mode)
            if actual_sims_completed >= 20:
                current_power = significant_count / actual_sims_completed
                if current_power == 0.0 and actual_sims_completed >= 50:
                    break
                elif current_power == 1.0 and actual_sims_completed >= 30:
                    break
    
    # Calculate final power
    power = significant_count / actual_sims_completed if actual_sims_completed > 0 else 0.0
    power = min(power, 1.0)
    return power

# Keep original function for backward compatibility
def simulate_clustered_binary_data_fixed_mice(baseline_prop, relative_effect, cluster_size, 
                                            median_icc, n_clusters=5, n_sims=1000):
    """
    Original non-vectorized version - kept for backward compatibility
    Uses parallel processing by default
    """
    return simulate_clustered_binary_data_fixed_mice_parallel(
        baseline_prop, relative_effect, cluster_size, median_icc, 
        n_clusters, n_sims, USE_PARALLEL, N_CORES
    )

def find_required_mice_for_effect_size(baseline_prop, relative_effect, cluster_size, 
                                     median_icc, target_power=0.8, max_mice=100, 
                                     n_sims=1000, use_parallel=True, n_cores=None):
    """
    Find minimum number of mice needed for given effect size using binary search
    
    Parameters:
    -----------
    baseline_prop : float
        Baseline disease proportion
    relative_effect : float
        Relative reduction in disease
    cluster_size : int
        Number of cells per mouse
    median_icc : float
        Intracluster correlation coefficient
    target_power : float
        Target statistical power (default 0.8)
    max_mice : int
        Maximum number of mice to test
    n_sims : int
        Number of simulations for power calculation
    use_parallel : bool
        Whether to use parallel processing
    n_cores : int
        Number of CPU cores to use
    """
    if n_cores is None:
        n_cores = N_CORES
        
    if relative_effect == 0:  # No effect, infinite mice needed
        return max_mice
    
    low = 2
    high = max_mice
    best_estimate = None
    
    # Binary search
    while low <= high:
        mid = (low + high) // 2
        
        try:
            power = simulate_clustered_binary_data_fixed_mice_parallel(
                baseline_prop=baseline_prop,
                relative_effect=relative_effect,
                cluster_size=cluster_size,
                median_icc=median_icc,
                n_clusters=mid,
                n_sims=n_sims,
                use_parallel=use_parallel,
                n_cores=n_cores
            )
            
            if abs(power - target_power) <= 0.05:  # Within 5% of target (tighter tolerance)
                return mid
            elif power < target_power:
                low = mid + 1
            else:
                best_estimate = mid
                high = mid - 1
                
        except Exception as e:
            # If simulation fails, try with fewer mice
            high = mid - 1
    
    return best_estimate if best_estimate is not None else max_mice

def run_power_analysis(icc_results, baseline_prop=BASELINE_PROP, max_mice=MAX_MICE_TESTED, 
                      n_sims=N_SIMULATIONS, relative_effects=RELATIVE_EFFECTS, 
                      cluster_sizes=CLUSTER_SIZES, target_power=TARGET_POWER,
                      use_parallel=USE_PARALLEL, n_cores=N_CORES):
    """
    Run complete power analysis with configurable parameters
    """
    
    # Get median ICC
    all_icc_values = []
    for dataset in icc_results.keys():
        for treatment in icc_results[dataset].keys():
            if treatment not in ['overall', 'Wildtype']:
                for outcome in icc_results[dataset][treatment].keys():
                    icc_val = icc_results[dataset][treatment][outcome]
                    if icc_val is not None and not np.isnan(icc_val):
                        all_icc_values.append(icc_val)

    median_icc = np.median(all_icc_values)
    print(f"Using median ICC: {median_icc:.4f}")
    print(f"Running {n_sims:,} simulations per analysis")
    print(f"Maximum mice tested per group: {max_mice}")
    if use_parallel:
        print(f"Using ProcessPoolExecutor with {n_cores} cores")
    else:
        print("Using sequential processing")
    
    print(f"Finding required mice for {target_power*100}% power across effect sizes")
    print(f"Testing {len(relative_effects)} effect sizes Ã— {len(cluster_sizes)} cell counts")

    # Run analysis
    results = []
    total_combinations = len(relative_effects) * len(cluster_sizes)

    with tqdm(total=total_combinations, desc="Finding required mice per effect size") as pbar:
        for cluster_size in cluster_sizes:
            for relative_effect in relative_effects:
                try:
                    required_mice = find_required_mice_for_effect_size(
                        baseline_prop=baseline_prop,
                        relative_effect=relative_effect,
                        cluster_size=cluster_size,
                        median_icc=median_icc,
                        target_power=target_power,
                        max_mice=max_mice,
                        n_sims=n_sims,
                        use_parallel=use_parallel,
                        n_cores=n_cores
                    )
                except Exception as e:
                    print(f"Error for effect {relative_effect*100}%, cluster {cluster_size}: {e}")
                    required_mice = max_mice
                
                results.append({
                    'relative_effect_pct': relative_effect * 100,
                    'cluster_size': cluster_size,
                    'required_mice_per_group': required_mice
                })
                
                pbar.update(1)

    # Convert to DataFrame
    mice_required_df = pd.DataFrame(results)
    
    # Create the plot
    fig, ax = plt.subplots(figsize=(9, 3))

    # Plot lines for each cell count
    colors = sns.color_palette("deep", len(cluster_sizes))
    for i, cluster_size in enumerate(cluster_sizes):
        cluster_data = mice_required_df[mice_required_df['cluster_size'] == cluster_size]
        
        ax.plot(cluster_data['relative_effect_pct'], cluster_data['required_mice_per_group'], 
                marker='o', linewidth=3, markersize=6, color=colors[i],
                label=f'{cluster_size} cells/mouse')

    ax.set_xlabel('Effect Size (% Reduction from Baseline)')
    ylabel_text = f'Mice Required per Group (for {target_power*100}% Power)'
    ax.set_ylabel(ylabel_text)
    title_text = f'Sample Size Requirements vs Effect Size\n({baseline_prop*100}% Baseline Disease Rate, ICC = {median_icc:.4f})'
    ax.set_title(title_text)
    ax.grid(True, alpha=0.3)
    ax.legend(title='Sampling Strategy')

    # Set axis limits
    ax.set_xlim(5, 80)
    ax.set_ylim(0, max_mice)

    # Add annotations
    parallel_text = f"ProcessPool ({n_cores} cores)" if use_parallel else "Sequential"
    target_text = f'Target: {target_power*100}% Power'
    baseline_text = f'Baseline: {baseline_prop*100}% diseased'
    sims_text = f'Sims: {n_sims:,}'
    mode_text = f'Mode: {parallel_text}'
    
    annotation_text = target_text + '\n' + baseline_text + '\n' + sims_text + '\n' + mode_text
    
    ax.text(0.02, 0.98, annotation_text, 
            transform=ax.transAxes, fontsize=10, verticalalignment='top',
            bbox=dict(boxstyle='round', facecolor='lightgray', alpha=0.8))

    plt.tight_layout()
    plt.show()

    print("Sample size analysis completed!")
    print("Results stored in 'mice_required_df' DataFrame")

    # Show key insights
    power_text = f"\nKey insights for {target_power*100}% power:"
    print(power_text)
    for cluster_size in cluster_sizes:
        cluster_data = mice_required_df[mice_required_df['cluster_size'] == cluster_size]
        print(f"\n{cluster_size} cells/mouse:")
        for effect in [10, 20, 30, 50]:
            effect_data = cluster_data[cluster_data['relative_effect_pct'] == effect]
            if len(effect_data) > 0:
                mice_needed = effect_data['required_mice_per_group'].iloc[0]
                if mice_needed == max_mice:
                    print(f"  {effect}% effect: >{max_mice} mice per group (underpowered)")
                else:
                    print(f"  {effect}% effect: {mice_needed} mice per group")

    # Calculate efficiency gains
    if len(cluster_sizes) >= 2:
        max_cluster = max(cluster_sizes)
        min_cluster = min(cluster_sizes)
        efficiency_text = f"\nEfficiency comparison (mice saved by using {max_cluster} vs {min_cluster} cells):"
        print(efficiency_text)
        for effect in [10, 20, 30, 50]:
            mice_min_data = mice_required_df[(mice_required_df['cluster_size'] == min_cluster) & 
                                           (mice_required_df['relative_effect_pct'] == effect)]
            mice_max_data = mice_required_df[(mice_required_df['cluster_size'] == max_cluster) & 
                                           (mice_required_df['relative_effect_pct'] == effect)]
            
            if len(mice_min_data) > 0 and len(mice_max_data) > 0:
                mice_min = mice_min_data['required_mice_per_group'].iloc[0]
                mice_max = mice_max_data['required_mice_per_group'].iloc[0]
                
                if mice_min == max_mice:
                    effect_text = f"  {effect}% effect: {min_cluster} cells/mouse requires >{max_mice} mice (vs {mice_max} for {max_cluster} cells)"
                    print(effect_text)
                else:
                    savings = mice_min - mice_max
                    savings_text = f"  {effect}% effect: {savings} fewer mice needed ({mice_min} â†’ {mice_max})"
                    print(savings_text)
    
    return mice_required_df, median_icc

# =============================================================================
# USAGE EXAMPLE (uncomment to run):
# =============================================================================
# Assuming you have icc_results from previous analysis:
mice_required_df, median_icc = run_power_analysis(icc_results)
```

```python
import numpy as np
from scipy import stats

def calculate_power_normal(baseline_mean, relative_effect, n_per_group, 
                          baseline_std=None, alpha=0.05, two_sided=True):
    """
    Calculate statistical power for detecting a relative effect between two groups
    assuming normal distributions and equal variances.
    
    Parameters:
    -----------
    baseline_mean : float
        Mean of the control/baseline group
    relative_effect : float
        Relative reduction from baseline (e.g., 0.2 = 20% reduction)
        Positive values = reduction, negative values = increase
    n_per_group : int
        Sample size per group
    baseline_std : float, optional
        Standard deviation. If None, assumes std = baseline_mean * 0.3 (30% CV)
    alpha : float, default 0.05
        Type I error rate (significance level)
    two_sided : bool, default True
        Whether to use two-sided test
        
    Returns:
    --------
    power : float
        Statistical power (probability of detecting the effect)
    effect_size_cohen_d : float
        Cohen's d effect size
    
    Examples:
    ---------
    # 20% reduction from baseline of 100, with 50 subjects per group
    power, d = calculate_power_normal(100, 0.2, 50)
    
    # 15% increase from baseline of 50, with custom std and one-sided test
    power, d = calculate_power_normal(50, -0.15, 30, baseline_std=12, two_sided=False)
    """
    
    # Calculate treatment mean
    treatment_mean = baseline_mean * (1 - relative_effect)
    
    # Estimate standard deviation if not provided (assume 30% coefficient of variation)
    if baseline_std is None:
        baseline_std = abs(baseline_mean) * 0.3
    
    # Calculate effect size (Cohen's d)
    mean_difference = abs(baseline_mean - treatment_mean)
    pooled_std = baseline_std  # Assuming equal variances
    cohens_d = mean_difference / pooled_std
    
    # Calculate standard error of the difference
    se_diff = pooled_std * np.sqrt(2 / n_per_group)
    
    # Calculate critical value
    if two_sided:
        critical_z = stats.norm.ppf(1 - alpha/2)
    else:
        critical_z = stats.norm.ppf(1 - alpha)
    
    # Calculate power
    z_beta = (mean_difference / se_diff) - critical_z
    power = stats.norm.cdf(z_beta)
    
    return power, cohens_d

def power_summary(baseline_mean, relative_effect, n_per_group, 
                 baseline_std=None, alpha=0.05, two_sided=True):
    """
    Calculate power and print a nice summary
    """
    power, cohens_d = calculate_power_normal(baseline_mean, relative_effect, n_per_group,
                                           baseline_std, alpha, two_sided)
    
    treatment_mean = baseline_mean * (1 - relative_effect)
    effect_direction = "reduction" if relative_effect > 0 else "increase"
    test_type = "two-sided" if two_sided else "one-sided"
    
    print(f"Power Analysis Summary:")
    print(f"=" * 40)
    print(f"Baseline mean:       {baseline_mean:.2f}")
    print(f"Treatment mean:      {treatment_mean:.2f}")
    print(f"Relative effect:     {relative_effect*100:.1f}% {effect_direction}")
    print(f"Sample size/group:   {n_per_group}")
    print(f"Standard deviation:  {baseline_std if baseline_std else baseline_mean*0.3:.2f}")
    print(f"Test type:           {test_type}")
    print(f"Alpha level:         {alpha}")
    print(f"")
    print(f"Results:")
    print(f"Cohen's d:           {cohens_d:.3f}")
    print(f"Statistical power:   {power:.3f} ({power*100:.1f}%)")
    
    # Interpret Cohen's d
    if cohens_d < 0.2:
        effect_interp = "very small"
    elif cohens_d < 0.5:
        effect_interp = "small"
    elif cohens_d < 0.8:
        effect_interp = "medium"
    else:
        effect_interp = "large"
    
    print(f"Effect size:         {effect_interp}")
    
    return power, cohens_d

# Quick examples
if __name__ == "__main__":
    print("Example 1: 20% reduction from baseline of 100")
    power_summary(0.25, 0.2, 36)
    
#     print("\n" + "="*50 + "\n")
    
#     print("Example 2: 15% increase with custom parameters")
#     power_summary(50, -0.15, 30, baseline_std=12, two_sided=False)
```

```python
import pandas as pd
import numpy as np
import statsmodels.api as sm
from statsmodels.genmod.families import Binomial, Gaussian

# Try different import paths for correlation structures (varies by statsmodels version)
try:
    from statsmodels.genmod.cov_struct import Exchangeable, Independence
except ImportError:
    try:
        from statsmodels.genmod.dependence import Exchangeable, Independence
    except ImportError:
        # Fallback for older versions
        from statsmodels.genmod import cov_struct
        Exchangeable = cov_struct.Exchangeable
        Independence = cov_struct.Independence

def gee_analysis(df, outcome_var, predictor_var, cluster_var, family='binomial', reference_level=None, fence_cutoff=1.7):
    """
    GEE analysis with both exchangeable and independence correlation structures
    
    Parameters:
    -----------
    df : pandas.DataFrame
        Input data
    outcome_var : str
        Name of outcome variable column
    predictor_var : str
        Name of main predictor variable
    cluster_var : str
        Name of clustering variable (e.g., 'filename' for mice)
    family : str, default='binomial'
        Family for GLM ('binomial' for binary, 'gaussian' for continuous)
    reference_level : str, optional
        Reference level for predictor (if None, uses first level alphabetically)
    fence_cutoff : float, default=1.7
        Multiplier for IQR to determine outlier fence (e.g., 1.5 for standard boxplot, 1.7 for slightly more conservative)
    
    Returns:
    --------
    results : dict
        Dictionary with 'exchangeable', 'independence', and 'exchangeable_no_outliers' results
    """
    
    # Create a copy to avoid modifying original data
    data = df.copy()
    
    # Remove rows with missing outcome
    data = data.dropna(subset=[outcome_var])
    
    # Set up family
    if family.lower() == 'binomial':
        model_family = Binomial()
    elif family.lower() == 'gaussian':
        model_family = Gaussian()
    else:
        raise ValueError(f"Family '{family}' not supported. Use 'binomial' or 'gaussian'")
    
    # Set reference level
    predictor_levels = data[predictor_var].unique()
    if reference_level is None:
        reference_level = sorted(predictor_levels)[0]
    elif reference_level not in predictor_levels:
        raise ValueError(f"Reference level '{reference_level}' not found in {predictor_var}")
    
    # Create dummy variables with specified reference
    treatment_dummies = pd.get_dummies(data[predictor_var], prefix='treatment', drop_first=False)
    ref_col = f'treatment_{reference_level}'
    if ref_col in treatment_dummies.columns:
        treatment_dummies = treatment_dummies.drop(columns=[ref_col])
    
    data = pd.concat([data, treatment_dummies], axis=1)
    treatment_cols = [col for col in data.columns if col.startswith('treatment_')]
    
    # Print number of mice per group
    mice_per_group = data.groupby(predictor_var)[cluster_var].nunique()
    print("Number of mice per group:")
    for group, count in mice_per_group.items():
        print(f"  {group}: {count}")
    print()
    
    # Calculate and print group means (first average within each mouse, then within group)
    mouse_means = data.groupby([cluster_var, predictor_var])[outcome_var].mean().reset_index()
    group_means = mouse_means.groupby(predictor_var)[outcome_var].mean()
    
    print("Group means (averaged within mouse, then within group):")
    for group, mean_val in group_means.items():
        print(f"  {group}: {mean_val:.4f}")
    print()
    
    # Create design matrix (add constant for intercept)
    X = sm.add_constant(data[treatment_cols])
    
    def format_pvalue(p):
        """Format p-value with at least 6 decimal places, extending if needed"""
        if p == 0:
            return "0.000000"
        
        # Start with 6 decimal places
        formatted = f"{p:.6f}"
        
        # If it ends with all zeros after decimal, keep extending until non-zero
        if '.' in formatted and formatted.split('.')[1] == '000000':
            for decimals in range(7, 16):  # Check up to 15 decimal places
                test_format = f"{p:.{decimals}f}"
                if not test_format.endswith('0'):
                    formatted = test_format
                    break
        
        return formatted
    
    def print_parameter_table(result, title):
        """Print formatted parameter estimates table"""
        print(f"=== {title} ===")
        print("Parameter estimates:")
        
        # Create a custom formatted table
        params_df = pd.DataFrame({
            'coef': result.params,
            'std err': result.bse,
            'z': result.tvalues,
            'P>|z|': result.pvalues,
            '[0.025': result.conf_int().iloc[:, 0],
            '0.975]': result.conf_int().iloc[:, 1]
        })
        
        # Format the dataframe for display
        print("=" * 95)
        print(f"{'':>25} {'coef':>12} {'std err':>10} {'z':>10} {'P>|z|':>18} {'[0.025':>12} {'0.975]':>12}")
        print("-" * 95)
        
        for idx, row in params_df.iterrows():
            pval_formatted = format_pvalue(row['P>|z|'])
            print(f"{idx:>25} {row['coef']:>12.4f} {row['std err']:>10.3f} {row['z']:>10.3f} {pval_formatted:>18} {row['[0.025']:>12.3f} {row['0.975]']:>12.3f}")
        
        print("=" * 95)
        print()
    
    # Identify and remove outlier mice using boxplot method
    def identify_outlier_mice(data, outcome_var, cluster_var, fence_cutoff):
        """Identify outlier mice using boxplot method with customizable fence"""
        # Calculate mean outcome for each mouse
        mouse_means = data.groupby(cluster_var)[outcome_var].mean()
        
        # Calculate IQR and fences
        Q1 = mouse_means.quantile(0.25)
        Q3 = mouse_means.quantile(0.75)
        IQR = Q3 - Q1
        
        lower_fence = Q1 - fence_cutoff * IQR
        upper_fence = Q3 + fence_cutoff * IQR
        
        # Identify outlier mice
        outlier_mice = mouse_means[(mouse_means < lower_fence) | (mouse_means > upper_fence)].index
        
        if len(outlier_mice) > 0:
            print(f"Outlier mice identified (using {fence_cutoff}*IQR rule): {list(outlier_mice)}")
            print(f"Mouse means - Q1: {Q1:.4f}, Q3: {Q3:.4f}, IQR: {IQR:.4f}")
            print(f"Fences - Lower: {lower_fence:.4f}, Upper: {upper_fence:.4f}")
            print()
        else:
            print(f"No outlier mice identified using {fence_cutoff}*IQR method")
            print()
        
        return outlier_mice
    
    # Identify outliers
    outlier_mice = identify_outlier_mice(data, outcome_var, cluster_var, fence_cutoff)
    
    # Create data without outliers
    data_no_outliers = data[~data[cluster_var].isin(outlier_mice)].copy()
    
    if len(outlier_mice) > 0:
        print(f"Data after removing outliers: {len(data_no_outliers)} observations from {data_no_outliers[cluster_var].nunique()} mice")
        print()
    
    # Fit three models
    results = {}
    
    # 1. Exchangeable correlation
    print_parameter_table(
        sm.GEE(endog=data[outcome_var], exog=X, groups=data[cluster_var], 
               family=model_family, cov_struct=Exchangeable()).fit(),
        "EXCHANGEABLE CORRELATION STRUCTURE"
    )
    results['exchangeable'] = sm.GEE(endog=data[outcome_var], exog=X, groups=data[cluster_var], 
                                   family=model_family, cov_struct=Exchangeable()).fit()
    
    # 2. Independence correlation  
    print_parameter_table(
        sm.GEE(endog=data[outcome_var], exog=X, groups=data[cluster_var], 
               family=model_family, cov_struct=Independence()).fit(),
        "INDEPENDENCE CORRELATION STRUCTURE"
    )
    results['independence'] = sm.GEE(endog=data[outcome_var], exog=X, groups=data[cluster_var], 
                                   family=model_family, cov_struct=Independence()).fit()
    
    # 3. Exchangeable without outliers (if any outliers were found)
    if len(outlier_mice) > 0:
        # Create design matrix for data without outliers
        X_no_outliers = sm.add_constant(data_no_outliers[treatment_cols])
        
        print_parameter_table(
            sm.GEE(endog=data_no_outliers[outcome_var], exog=X_no_outliers, 
                   groups=data_no_outliers[cluster_var], family=model_family, 
                   cov_struct=Exchangeable()).fit(),
            "EXCHANGEABLE WITHOUT OUTLIERS"
        )
        results['exchangeable_no_outliers'] = sm.GEE(endog=data_no_outliers[outcome_var], 
                                                    exog=X_no_outliers, 
                                                    groups=data_no_outliers[cluster_var], 
                                                    family=model_family, 
                                                    cov_struct=Exchangeable()).fit()
    else:
        print("=== EXCHANGEABLE WITHOUT OUTLIERS ===")
        print("No outliers detected - same as exchangeable model above")
        print()
        results['exchangeable_no_outliers'] = results['exchangeable']
    
    # After fitting all models, create comparison dataframe
    print("=== COMPARISON TABLE ===")
    
    # Extract results for comparison
    exch_result = results['exchangeable']
    indep_result = results['independence']
    exch_no_outliers_result = results['exchangeable_no_outliers']
    
    # Create comparison dataframe
    comparison_data = []
    
    for param_name in exch_result.params.index:
        if param_name != 'const':  # Skip intercept
            # Clean up treatment name (remove 'treatment_' prefix)
            clean_name = param_name.replace('treatment_', '') if param_name.startswith('treatment_') else param_name
            
            comparison_data.append({
                'treatment_group': clean_name,
                'coef': exch_result.params[param_name],
                'independence_pvalue': indep_result.pvalues[param_name],
                'exchangeable_pvalue': exch_result.pvalues[param_name],
                'exchangeable_no_outliers_pvalue': exch_no_outliers_result.pvalues[param_name]
            })
    
    comparison_df = pd.DataFrame(comparison_data)
    
    # Format p-values in scientific notation
    comparison_df['independence_pvalue'] = comparison_df['independence_pvalue'].apply(lambda x: f"{x:.2e}")
    comparison_df['exchangeable_pvalue'] = comparison_df['exchangeable_pvalue'].apply(lambda x: f"{x:.2e}")
    comparison_df['exchangeable_no_outliers_pvalue'] = comparison_df['exchangeable_no_outliers_pvalue'].apply(lambda x: f"{x:.2e}")
    
    # Round coefficient to 4 decimal places
    comparison_df['coef'] = comparison_df['coef'].round(4)
    
    display(comparison_df)
    print()
    
    return results
```
