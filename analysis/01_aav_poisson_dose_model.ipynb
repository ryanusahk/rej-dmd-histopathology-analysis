```python
from scipy import stats
from scipy.stats import norm
```

```python
class mix_norm(stats.rv_continuous):
    def _pdf(self, x, m1, s1, m2, s2, r):
        n1 = (np.exp(-0.5*((x-m1)/s1)**2))/(s1*np.sqrt(2*np.pi))
        n2 = (np.exp(-0.5*((x-m2)/s2)**2))/(s2*np.sqrt(2*np.pi))
        return (r*n1+n2)/(r+1)
```

```python
import numpy as np
from scipy import stats

# Your data
doses = np.array([5e9, 5e10, 5e11])
mois = np.array([0.273889, 2.768984, 18.560859])

# Fit linear relationship through origin: MOI = m * Dose
slope = np.sum(doses * mois) / np.sum(doses**2)

print(f"Linear fit through origin:")
print(f"MOI = {slope:.2e} * Dose")

# Function to predict MOI from dose
def predict_moi(dose):
    return slope * dose

# Function to predict dose from target MOI
def predict_dose(target_moi):
    return target_moi / slope

# Find dose for 90% filled (10% empty, MOI â‰ˆ 2.3)
target_moi = -np.log(0.10)  # MOI for 10% empty
required_dose = predict_dose(target_moi)

print(f"\nFor 90% filled wells:")
print(f"Target MOI: {target_moi:.3f}")
print(f"Required dose: {required_dose:.2e}")

# Calculate RÂ² for the through-origin fit
predicted_mois = predict_moi(doses)
ss_res = np.sum((mois - predicted_mois)**2)
ss_tot = np.sum(mois**2)
r_squared = 1 - (ss_res / ss_tot)
print(f"RÂ² (through origin): {r_squared:.3f}")

# Verify the fit
print(f"\nFit verification:")
for i, dose in enumerate(doses):
    predicted = predict_moi(dose)
    print(f"Dose {dose:.0e}: Actual MOI={mois[i]:.3f}, Predicted MOI={predicted:.3f}")
```

```python
def pTripleInfectionAnalytical(MOI, ratioA=1/3, ratioB=1/3):
    """
    Analytical solution for probability of triple infection (â‰¥1 of each type).
    """
    ratioC = 1 - ratioA - ratioB
    
    # P(0 types) = P(no infections)
    p_zero_types = sp.stats.poisson.pmf(0, MOI)
    
    # P(1 type only)
    p_only_A = np.exp(-MOI) * (np.exp(MOI * ratioA) - 1)  # = exp(-MOI*(1-ratioA)) - exp(-MOI)
    p_only_B = np.exp(-MOI) * (np.exp(MOI * ratioB) - 1)
    p_only_C = np.exp(-MOI) * (np.exp(MOI * ratioC) - 1)
    p_one_type = p_only_A + p_only_B + p_only_C
    
    # P(2 types only)
    p_AB_no_C = np.exp(-MOI) * (np.exp(MOI * (ratioA + ratioB)) - np.exp(MOI * ratioA) - np.exp(MOI * ratioB) + 1)
    p_AC_no_B = np.exp(-MOI) * (np.exp(MOI * (ratioA + ratioC)) - np.exp(MOI * ratioA) - np.exp(MOI * ratioC) + 1)
    p_BC_no_A = np.exp(-MOI) * (np.exp(MOI * (ratioB + ratioC)) - np.exp(MOI * ratioB) - np.exp(MOI * ratioC) + 1)
    p_two_types = p_AB_no_C + p_AC_no_B + p_BC_no_A
    
    # Triple infection = complement
    return 1 - p_zero_types - p_one_type - p_two_types

# Much faster calculation
ysss_analytical = []
for m in tqdm(xs):
    ysss_analytical.append(100 * pTripleInfectionAnalytical(m))
```

```python
import pandas as pd
import numpy as np
from scipy import interpolate

# Create filled percentages from 0 to 99
filled_percents = np.arange(0, 100)

# Create interpolation functions for dual and triple vector systems
# Curves are in 0-100 range
dual_interp = interpolate.interp1d(ys, xs, kind='linear', bounds_error=False, fill_value='extrapolate')
triple_interp = interpolate.interp1d(ysss_analytical, xs, kind='linear', bounds_error=False, fill_value='extrapolate')

# Calculate doses for each vector type
results = []
for pct in filled_percents:
   if pct == 0:
       # Handle 0% case
       single_moi = dual_moi = triple_moi = 0
   else:
       # Single vector: use infectivity approach (Poisson)
       filled_fraction = pct / 100
       if filled_fraction >= 0.999999:  # Avoid log(0)
           filled_fraction = 0.999999
       single_moi = -np.log(1 - filled_fraction)
       
       # Dual and triple: use interpolation from your curves (0-100 scale)
       dual_moi = dual_interp(pct)
       triple_moi = triple_interp(pct)
   
   # Convert MOI to dose using infectivity
   single_dose = single_moi / infectivity
   dual_dose = dual_moi / infectivity
   triple_dose = triple_moi / infectivity
   
   results.append({
       'filled_percent': pct,
       'single_vector_moi': f"{single_moi:.3f}",
       'dual_vector_moi': f"{dual_moi:.3f}",
       'triple_vector_moi': f"{triple_moi:.3f}",
       'single_vector_dose': f"{single_dose:.2e}",
       'dual_vector_dose': f"{dual_dose:.2e}",
       'triple_vector_dose': f"{triple_dose:.2e}",
       'dual_vs_single_ratio': dual_dose / single_dose if single_dose > 0 else np.nan,
       'triple_vs_single_ratio': triple_dose / single_dose if single_dose > 0 else np.nan
   })

# Create DataFrame and export
dose_df = pd.DataFrame(results)
dose_df.to_csv('vector_dose_requirements.csv', index=False)

print("CSV exported! Here's a preview:")
print(dose_df.head(10))
```
